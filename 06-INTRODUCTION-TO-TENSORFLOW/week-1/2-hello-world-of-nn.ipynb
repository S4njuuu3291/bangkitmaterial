## Menemukan Pola

Oke, berikut adalah angkanya. Ada sebuah formula yang memetakan X ke Y. Dapatkah Anda menemukannya? Luangkan waktu sejenak.

**Jawabannya adalah**: 
$$ Y = 2X - 1 $$

Jadi, setiap kali Anda melihat Y, itu dua kali lipat dari X yang sesuai, dikurangi 1. Jika Anda berhasil menemukan ini sendiri, **selamat!** 

### Bagaimana Anda Melakukannya?

Bagaimana cara Anda menemukannya? Mungkin Anda melihat bahwa Y meningkat 2 setiap kali X meningkat 1. Jadi, kemungkinan bentuknya seperti:
$$ Y = 2X \pm \text{sesuatu} $$

Ketika Anda melihat $ X = 0 $ dan $ Y = -1 $, Anda berpikir, "Hey, sesuatu itu adalah -1." Jadi, jawabannya mungkin:
$$ Y = 2X - 1 $$

Anda mungkin mencoba hal ini dengan beberapa nilai lain dan melihat bahwa itu cocok.

### Apa itu Jaringan Saraf?

Jaringan saraf, pada dasarnya, adalah sekumpulan fungsi yang dapat belajar pola. Jangan khawatir jika ada banyak konsep baru di sini; semuanya akan menjadi jelas dengan cepat saat Anda mempelajarinya.

Jaringan saraf yang paling sederhana adalah yang hanya memiliki satu neuron. Dan itulah yang dilakukan oleh baris kode ini. Dalam Keras, Anda menggunakan kata **Dense** untuk mendefinisikan lapisan neuron yang terhubung.
