# Rekomendasi Produk pada Website E-Commerce

Ketika kamu mengunjungi sebuah website e-commerce dan sedang melihat produk tertentu, misalnya sebuah buku, website tersebut mungkin akan menunjukkan rekomendasi seperti, "Berikut adalah buku-buku lain yang mirip dengan ini" atau jika kamu sedang melihat film, mungkin akan muncul, "Berikut adalah film-film lain yang mirip dengan ini."

## Bagaimana Website Melakukan Itu?

Bagaimana website bisa memberikan rekomendasi produk serupa saat kamu melihat satu produk? Ternyata, **Collaborative Filtering** adalah salah satu algoritma yang digunakan untuk menemukan item yang berkaitan.

### Collaborative Filtering

Dalam **Collaborative Filtering**, kita mempelajari **fitur** $x^{(i)}$ untuk setiap item $i$, seperti film atau produk lain yang direkomendasikan kepada pengguna. Sebelumnya, kita menggunakan contoh hipotesis bahwa fitur ini mewakili seberapa besar sebuah film termasuk genre romantis atau aksi. Namun dalam praktiknya, saat algoritma digunakan untuk mempelajari fitur $x^{(i)}$ secara otomatis, fitur-fitur individual seperti $x_1$, $x_2$, dan $x_3$ sulit diinterpretasikan.

> **Catatan:** Fitur-fitur ini tetap menyampaikan informasi tentang karakteristik sebuah film, walaupun sulit diartikan secara individu.



Jika kamu ingin menemukan item lain yang mirip dengan item $i$, kamu bisa mencari item $k$ dengan fitur $x^{(k)}$ yang mirip dengan $x^{(i)}$. Secara khusus, kita bisa menghitung jarak antara dua vektor fitur $x^{(k)}$ dan $x^{(i)}$ sebagai berikut:

$$
\sum_{l=1}^{n} \left( x^{(k)}_l - x^{(i)}_l \right)^2
$$

Rumus ini menghitung **jarak kuadrat** antara dua vektor fitur, dan kita bisa mencari item-item dengan jarak terdekat untuk menemukan produk yang mirip.

### Rekomendasi Berdasarkan Fitur

Jika kita tidak hanya mencari satu produk, tetapi 5 hingga 10 produk dengan vektor fitur yang paling mirip, maka kita bisa menampilkan 5 hingga 10 item yang berkaitan dengan item $i$. Teknik ini sangat berguna jika kita ingin menampilkan produk terkait kepada pengguna.

## Keterbatasan Collaborative Filtering

Walaupun algoritma ini sangat berguna, **Collaborative Filtering** memiliki beberapa keterbatasan, di antaranya:



### 1. Cold Start Problem

Algoritma ini tidak terlalu efektif dalam menangani **cold start problem**. Misalnya, jika ada produk baru di katalog (seperti film baru yang baru dirilis) dan hanya sedikit pengguna yang memberikan rating, bagaimana cara kita merekomendasikan produk tersebut? Hal yang sama berlaku untuk pengguna baru yang hanya memberikan rating pada sedikit item. 

### 2. Keterbatasan Informasi Tambahan

Algoritma ini juga tidak secara alami memanfaatkan **informasi tambahan** tentang item atau pengguna. Misalnya, untuk sebuah film, kita mungkin memiliki informasi tentang:

- Genre
- Aktor
- Studio produksi
- Budget

Sedangkan untuk pengguna, kita mungkin tahu hal-hal seperti:

- Usia
- Jenis kelamin
- Lokasi

Bahkan informasi seperti **alamat IP** atau **jenis perangkat** (mobile atau desktop) bisa memberikan petunjuk tentang preferensi pengguna. Hal ini dikenal sebagai **side information** dan bisa sangat membantu dalam membuat prediksi yang lebih akurat.

## Solusi untuk Keterbatasan Collaborative Filtering

Untuk mengatasi beberapa keterbatasan ini, kita bisa menggunakan algoritma lain seperti **Content-Based Filtering**, yang akan kita bahas selanjutnya. Algoritma ini digunakan secara luas dalam aplikasi komersial modern dan dapat memberikan hasil yang lebih baik dalam situasi di mana Collaborative Filtering memiliki kelemahan.
