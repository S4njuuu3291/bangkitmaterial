# Recommender Systems dan Optimisasi Efisiensi

Sistem rekomendasi saat ini sering kali harus memilih sejumlah item untuk direkomendasikan dari katalog yang sangat besarbisa mencapai ribuan, jutaan, hingga puluhan juta item. Bagaimana cara melakukannya secara efisien? Mari kita lihat lebih dalam. Dalam neural network yang digunakan untuk membuat prediksi tentang bagaimana seorang pengguna mungkin memberi rating pada sebuah item, mungkin terdapat ribuan film di platform streaming atau jutaan iklan di sistem iklan, atau jutaan lagu di platform streaming musik yang harus dipilih.

Ketika seorang pengguna membuka website Anda, mereka memiliki fitur **Xu**. Namun, jika Anda harus memproses jutaan item melalui neural network untuk menentukan produk yang direkomendasikan, menjalankan inferensi neural network jutaan kali setiap kali seorang pengguna datang menjadi tidak mungkin secara komputasional.

Banyak sistem rekomendasi skala besar diimplementasikan dalam dua langkah: **retrieval** dan **ranking**.

## Langkah Retrieval
Pada langkah retrieval, sistem akan menghasilkan daftar besar kandidat item yang mungkin sesuai. Pada tahap ini, tidak masalah jika ada banyak item yang mungkin tidak disukai pengguna. 

Contohnya:
- Untuk setiap 10 film terakhir yang ditonton pengguna, kita bisa mencari 10 film yang paling mirip.
- Jika pengguna menonton film dengan vektor **VIM**, kita bisa menemukan film dengan vektor **VKM** yang mirip. Pencarian film yang mirip ini bisa dipra-komputasi dan hasilnya bisa diambil dari tabel lookup.
  
Contoh lain, kita bisa menambahkan top 10 film dari tiga genre yang sering ditonton pengguna, seperti **romance**, **comedy**, dan **historical drama**. Kita juga bisa menambahkan top 20 film dari negara pengguna. Langkah retrieval ini dapat dilakukan dengan sangat cepat, dan Anda mungkin akan mendapatkan sekitar 100 atau lebih kandidat film.

## Langkah Ranking
Setelah langkah retrieval, kita melanjutkan ke **ranking**. Pada langkah ini, item-item dari retrieval akan diranking menggunakan model yang telah dipelajari. Artinya, kita akan mengumpankan fitur pengguna dan fitur item ke neural network untuk menghitung prediksi rating setiap pasangan pengguna-item. Berdasarkan hasil ini, kita bisa memilih item dengan rating tertinggi untuk ditampilkan kepada pengguna.

Optimasi tambahan adalah jika Anda sudah memiliki **VM** untuk semua film sebelumnya, maka Anda hanya perlu melakukan inferensi sekali untuk menghitung **VU**. Kemudian, Anda bisa menghitung produk dalam antara **VU** dan **VM** dari film-film yang diambil dari retrieval.

### Trade-off Efisiensi dan Akurasi
Salah satu keputusan penting adalah menentukan berapa banyak item yang ingin Anda ambil selama retrieval untuk diteruskan ke langkah ranking. Mengambil lebih banyak item cenderung meningkatkan performa, namun algoritma akan lebih lambat. Oleh karena itu, Anda perlu mengoptimalkan trade-off ini.

Untuk mengetahui seberapa banyak item yang harus diambil, disarankan melakukan eksperimen offline untuk melihat seberapa besar tambahan item yang diambil berpengaruh terhadap relevansi rekomendasi. Misalnya, jika estimasi rating atau probabilitas untuk item yang diambil jauh lebih tinggi saat mengambil 500 item dibandingkan 100 item, maka mungkin lebih baik mengambil lebih banyak item meskipun algoritma menjadi sedikit lebih lambat.

Dengan adanya langkah retrieval dan ranking yang terpisah, sistem rekomendasi modern dapat memberikan hasil yang cepat sekaligus akurat.

## Etika dalam Recommender Systems
Sistem rekomendasi memiliki dampak besar secara komersial, namun sayangnya, beberapa sistem ini juga dapat menyebabkan kerugian. Oleh karena itu, ketika Anda membangun sistem rekomendasi, penting untuk mempertimbangkan pendekatan etis agar sistem yang Anda bangun dapat bermanfaat bagi pengguna, masyarakat, perusahaan, dan diri Anda sendiri. Mari kita bahas masalah etika dalam sistem rekomendasi pada materi selanjutnya.
