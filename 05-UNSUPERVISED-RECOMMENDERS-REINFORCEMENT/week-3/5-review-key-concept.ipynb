# Pemrograman Pembelajaran Penguatan: Tinjauan Konsep Kunci

Kami telah mengembangkan sebuah **reinforcement learning** formalism menggunakan contoh rover Mars dengan enam keadaan. Mari kita tinjau konsep-konsep kunci dan lihat bagaimana rangkaian konsep ini dapat digunakan untuk aplikasi lain. 

## Konsep Utama dalam Pembelajaran Penguatan

Beberapa konsep yang telah kita diskusikan adalah:
- **States** dari masalah pembelajaran penguatan
- **Set of actions**
- **Rewards**
- **Discount factor**

Kemudian, bagaimana **rewards** dan **discount factor** secara keseluruhan digunakan untuk menghitung **return**, dan akhirnya, sebuah **policy** yang berfungsi untuk membantu Anda memilih aksi agar memaksimalkan return.

### Contoh Rover Mars

Untuk contoh rover Mars, kami memiliki enam keadaan yang kami beri nomor 1-6, dan aksinya adalah bergerak ke kiri atau ke kanan. 
- **Rewards**: 
  - 100 untuk keadaan paling kiri 
  - 40 untuk keadaan paling kanan 
  - Nol di antara keduanya 
- **Discount factor**: 0.5 

Return dihitung menggunakan formula tertentu dan kami dapat memiliki kebijakan \( \pi \) yang menggambarkan aksi tergantung pada keadaan yang Anda hadapi.

### Aplikasi Lain: Helikopter Otonom

Formalism yang sama ini (states, actions, rewards, dll.) dapat digunakan untuk banyak aplikasi lain. Ambil contoh masalah **helikopter otonom**. 
- **State**: Sekumpulan posisi, orientasi, dan kecepatan helikopter yang mungkin.
- **Actions**: Cara-cara untuk menggerakkan kendali helikopter.
- **Rewards**: +1 jika terbang dengan baik, -1000 jika jatuh atau mengalami kerusakan. 
- **Discount factor**: Angka sedikit kurang dari satu, misalnya 0.99. 

Berdasarkan rewards dan discount factor, Anda menghitung return menggunakan formula yang sama. 

### Contoh Lain: Permainan Catur

Berikut satu contoh lagi, yaitu bermain catur. 
- **State**: Posisi semua bidak di papan.
- **Actions**: Langkah legal yang mungkin dalam permainan.
- **Rewards**: 
  - +1 jika menang 
  - -1 jika kalah 
  - 0 jika seri. 

Biasanya, discount factor yang sangat mendekati satu digunakan, seperti 0.99 atau bahkan 0.995. 

## Markov Decision Process (MDP)

Formalism aplikasi pembelajaran penguatan ini sebenarnya memiliki nama, yaitu **Markov Decision Process (MDP)**. Istilah **Markov** dalam MDP merujuk pada fakta bahwa masa depan hanya bergantung pada keadaan saat ini, bukan pada apa pun yang mungkin terjadi sebelum mencapai keadaan saat ini. 

### Pemahaman MDP

Dalam proses keputusan Markov:
- **Robot atau agen** yang ingin kita kendalikan.
- Kita memilih **actions** \( a \).
- Berdasarkan aksi tersebut, sesuatu akan terjadi di dunia atau lingkungan, seperti posisi kita yang berubah atau kita melakukan sampel batuan.

Cara kita memilih aksi \( a \) adalah dengan menggunakan **policy** \( \pi \). Berdasarkan apa yang terjadi, kita akan mengamati keadaan dan rewards yang kita peroleh.
