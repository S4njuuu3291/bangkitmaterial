## Mengubah Nilai

Sekarang, mari kita mulai bersenang-senang. Mari kita ubah beberapa nilai dan lihat bagaimana ini mempengaruhi hasil. Saya akan memperbarui hadiah terminal kanan menjadi nilai yang jauh lebih kecil, yaitu hanya 10. Jika saya menjalankan kembali kode ini, lihat bagaimana Q(S,A) berubah dan sekarang berpikir bahwa jika Anda berada di keadaan 5, maka jika Anda pergi ke kiri dan berperilaku optimal, Anda akan mendapatkan 6.25. Sementara itu, jika Anda pergi ke kanan dan juga berperilaku optimal, Anda hanya akan mendapatkan pengembalian 5.

Sekarang, ketika hadiah di kanan sangat kecil, yaitu hanya 10, bahkan ketika Anda sangat dekat, lebih baik Anda pergi ke kiri. Faktanya, kebijakan otomatis sekarang adalah untuk pergi ke kiri dari setiap keadaan. Mari kita lakukan beberapa perubahan lainnya. Saya akan mengubah hadiah terminal kanan kembali ke 40, tetapi saya akan mengubah faktor diskonto menjadi 0.9. Dengan faktor diskonto yang lebih dekat ke satu, ini membuat Mars Rover kurang tidak sabar dan bersedia untuk menunggu lebih lama untuk mendapatkan hadiah yang lebih tinggi. Ini karena hadiah di masa depan tidak dikalikan dengan 0.5 ke pangkat tinggi, melainkan dikalikan dengan 0.9 ke pangkat tinggi. Jadi, ini bersedia lebih sabar karena hadiah di masa depan tidak didiskontokan atau dikalikan dengan angka yang sekecil saat diskonto adalah 0.5.
