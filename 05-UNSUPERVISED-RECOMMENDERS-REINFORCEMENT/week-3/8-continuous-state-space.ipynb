# Pengenalan Ruang Status Kontinu dalam Kontrol Robotik

Banyak aplikasi kontrol robotik, termasuk aplikasi **lunar lander** yang Anda kerjakan di lab praktik, memiliki ruang status kontinu. Mari kita lihat apa arti itu dan bagaimana kita dapat menggeneralisasi konsep yang telah kita bicarakan ke dalam ruang status kontinu ini.

## Contoh Rover Mars

Sebagai contoh yang disederhanakan, saya menggunakan **Mars rover**. Dalam contoh ini, saya menggunakan sekumpulan status diskrit, yang berarti **Mars rover** tersebut hanya bisa berada dalam salah satu dari enam posisi yang mungkin. Namun, sebagian besar robot dapat berada dalam lebih dari enam posisi diskrit; sebaliknya, mereka dapat berada dalam jumlah posisi nilai kontinu yang sangat besar. 

Misalnya, jika **Mars rover** bisa berada di mana saja di sepanjang garis, maka posisinya dapat diindikasikan oleh angka yang berkisar dari **0 hingga 6 kilometer**, di mana angka mana pun di antara itu adalah valid. Ini adalah contoh ruang status kontinu, karena posisi dapat diwakili oleh angka seperti **2,7 kilometer** atau **4,8 kilometer**, atau angka lain di antara nol dan enam.

## Contoh Kontrol Mobil atau Truk

Mari kita lihat contoh lain. Saya akan menggunakan aplikasi kontrol **mobil** atau **truk**. Berikut adalah gambar mainan truk yang dimiliki oleh putri saya. Jika Anda sedang membangun **mobil otonom** atau **truk otonom** dan ingin mengontrolnya agar berjalan dengan lancar, maka status truk ini mungkin mencakup beberapa angka seperti:

- **x position** (posisi x)
- **y position** (posisi y)
- **orientation** (orientasi)

Asumsikan truk tetap berada di tanah; Anda mungkin tidak perlu khawatir tentang seberapa tinggi truk tersebut. Status ini akan mencakup x, y, dan sudut **Theta**, serta mungkin kecepatannya di arah x (disebut **x dot**), yang menunjukkan seberapa cepat koordinat x ini berubah, **y dot** untuk perubahan koordinat y, dan akhirnya **Theta dot**, yang menunjukkan seberapa cepat sudut mobil berubah. 

Untuk truk atau mobil, status mungkin terdiri dari lebih dari satu angka, seperti berapa kilometer truk ini berada di sepanjang garis ini, tetapi dapat mencakup enam angka, yaitu:

1. x position (posisi x)
2. y position (posisi y)
3. orientation (orientasi) — dinyatakan dengan huruf Yunani **Theta**
4. velocity in the x-direction (kecepatan di arah x) — dinyatakan dengan **x dot**
5. velocity in the y-direction (kecepatan di arah y) — dinyatakan dengan **y dot**
6. angular velocity (kecepatan sudut) — dinyatakan dengan **Theta dot**

Di mana semua angka ini dapat mengambil nilai apa pun dalam rentang yang valid. Misalnya, **Theta** seharusnya berkisar antara **0 hingga 360 derajat**.

## Contoh Kontrol Helikopter Otonom

Bagaimana jika Anda membangun algoritma **reinforcement learning** untuk mengontrol **helikopter otonom**? Bagaimana Anda akan menggambarkan posisi helikopter? Untuk ilustrasi, saya memiliki sebuah helikopter mainan kecil. Posisi helikopter akan mencakup:

- **x position** (seberapa jauh utara atau selatan helikopter)
- **y position** (seberapa jauh di sumbu timur-barat helikopter)
- **z** (ketinggian helikopter di atas tanah)

Namun, selain posisi, helikopter juga memiliki orientasi. Secara konvensional, salah satu cara untuk menangkap orientasinya adalah dengan tiga angka tambahan, yang meliputi:

- **roll** (apakah helikopter berguling ke kiri atau kanan)
- **pitch** (apakah helikopter miring ke depan atau belakang)
- **yaw** (orientasi kompas — apakah helikopter menghadap utara, timur, selatan, atau barat)

### Ringkasan Status Helikopter

Dengan demikian, status helikopter mencakup:

- Posisi x
- Posisi y
- Ketinggian z
- Roll (dinyatakan dengan huruf Yunani **Phi**)
- Pitch (dinyatakan dengan huruf Yunani **Theta**)
- Yaw (dinyatakan dengan huruf Yunani **Omega**)

Namun, untuk mengontrol helikopter, kita juga perlu mengetahui kecepatan di arah x, arah y, dan arah z, serta laju perputaran (kecepatan sudut). Seberapa cepat roll, pitch, dan yaw berubah?

## Kesimpulan

Ini adalah daftar 12 angka yang digunakan sebagai input untuk suatu kebijakan (policy), dan tugas kebijakan tersebut adalah melihat 12 angka ini dan memutuskan tindakan yang sesuai yang harus diambil oleh helikopter. Jadi, dalam setiap masalah **reinforcement learning** dengan status kontinu atau **Markov decision process** (MDP) kontinu, status masalah tidak hanya satu dari sejumlah kecil nilai diskrit yang mungkin, seperti angka dari 1 hingga 6. Sebaliknya, ia merupakan vektor angka, di mana masing-masing bisa mengambil banyak nilai.
