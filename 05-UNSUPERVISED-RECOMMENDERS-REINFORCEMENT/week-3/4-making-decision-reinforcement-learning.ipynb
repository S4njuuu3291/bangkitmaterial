## Policy dalam Pembelajaran Penguatan

Dalam pembelajaran penguatan, tujuan kita adalah untuk menghasilkan fungsi yang disebut **policy** (Pi), yang tugasnya adalah menerima input dalam bentuk keadaan (state) **s** dan memetakan ke tindakan (action) **a** yang ingin kita ambil. 

Sebagai contoh, untuk policy ini di bawah, policy ini akan mengatakan bahwa jika Anda berada dalam state 2, maka ia memetakan kita ke tindakan ke kiri. Jika Anda berada dalam state 3, policy mengatakan untuk pergi ke kiri. Jika Anda berada dalam state 4, juga pergi ke kiri, dan jika Anda berada dalam state 5, pergi ke kanan. 

Pi yang diterapkan pada state S memberi tahu kita tindakan apa yang ingin diambil dalam state tersebut. 

## Tujuan Pembelajaran Penguatan

Tujuan dari pembelajaran penguatan adalah menemukan policy **Pi** atau **Pi(s)** yang memberi tahu Anda tindakan apa yang harus diambil di setiap state untuk memaksimalkan **return**. 

Omong-omong, saya tidak tahu apakah istilah **policy** adalah istilah yang paling deskriptif untuk apa itu Pi, tetapi ini adalah salah satu istilah yang telah menjadi standar dalam pembelajaran penguatan. Mungkin menyebut Pi sebagai **controller** alih-alih policy akan menjadi terminologi yang lebih natural, tetapi policy adalah apa yang sekarang disebut oleh semua orang dalam pembelajaran penguatan.
