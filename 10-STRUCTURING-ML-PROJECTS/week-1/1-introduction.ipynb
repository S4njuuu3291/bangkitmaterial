## Apa itu Machine Learning Strategy?

Mari kita mulai dengan sebuah contoh motivasi. Misalkan Anda sedang bekerja pada proyek *cat cost file*. Setelah beberapa waktu bekerja, Anda berhasil mencapai akurasi 90% untuk sistem Anda. Namun, hasil ini mungkin belum cukup baik untuk aplikasi Anda. 

Anda mungkin memiliki banyak ide untuk meningkatkan sistem ini, seperti:

- Mengumpulkan lebih banyak **data pelatihan**
- Mengambil gambar kucing dengan pose yang lebih beragam atau contoh negatif yang lebih beragam
- Melatih algoritma lebih lama menggunakan **gradient descent**
- Mencoba **algoritma optimasi** lain, seperti **Adam optimizer**
- Mencoba **jaringan neural** yang lebih besar atau lebih kecil
- Menggunakan **dropout** atau **L2 regularization**
- Mengubah **arsitektur jaringan**, seperti mengganti **activation function**, mengubah jumlah **hidden units**, dan sebagainya

Saat mencoba meningkatkan sistem *deep learning*, biasanya ada banyak ide atau hal yang bisa dicoba. Namun, jika pilihan Anda kurang tepat, bisa saja Anda menghabiskan enam bulan mengejar satu arah, hanya untuk menyadari bahwa arah itu tidak memberikan hasil yang diharapkan.

# Tantangan dalam Pengembangan Sistem Machine Learning

Salah satu tantangan dalam membangun sistem **machine learning** adalah terdapat begitu banyak hal yang bisa dicoba, dan begitu banyak hal yang bisa diubah, termasuk banyaknya **hyperparameter** yang bisa di-*tune*.

> **Transkrip**:
> Salah satu hal yang saya perhatikan pada orang-orang yang paling efektif dalam **machine learning** adalah mereka sangat jelas tentang apa yang perlu di-*tune* untuk mencapai efek tertentu. Proses ini kita sebut **orthogonalization**. Izinkan saya menjelaskan maksud saya.

### Contoh Televisi Lama dengan Banyak Knob

Bayangkan sebuah televisi model lama dengan banyak **knob** yang bisa di-*tune* untuk mengatur gambar dengan berbagai cara.

> **Transkrip**:
> Pada TV lama ini, mungkin ada satu knob untuk mengatur ketinggian gambar secara vertikal, satu lagi untuk lebarnya, satu lagi untuk membuat gambar trapezoidal, dan seterusnya. Desainer TV menghabiskan banyak waktu membangun rangkaian untuk memastikan setiap knob memiliki fungsi yang jelas. Misalnya, satu knob untuk mengatur ini, knob lain untuk mengatur itu, dan sebagainya.

Jika Anda memiliki knob yang mengatur berbagai aspek gambar secara bersamaan, maka akan sangat sulit untuk mengatur gambar agar berada di posisi yang diinginkan. **Orthogonalization** dalam konteks ini berarti setiap knob diatur agar hanya memengaruhi satu aspek tertentu, sehingga lebih mudah menyesuaikan gambar sesuai keinginan.



### Orthogonalization dalam Mengemudi Mobil

Contoh lain adalah saat mengemudi mobil: terdapat tiga kontrol utama - **kemudi** untuk arah kiri atau kanan, **gas**, dan **rem** untuk mengatur kecepatan. Setiap kontrol ini memiliki fungsi yang terpisah dan mudah dipahami.

> **Transkrip**:
> Bayangkan jika sebuah mobil dirancang dengan joystick yang menggabungkan berbagai kontrol, misalnya mengatur kemudi dan kecepatan secara bersamaan. Hal ini akan menyulitkan pengemudi untuk mengatur mobil pada kecepatan dan arah yang diinginkan.

**Orthogonalization** berarti memiliki kontrol yang terpisah untuk setiap aspek, sehingga memudahkan pengaturan kendaraan sesuai keinginan. Lalu, bagaimana ini berkaitan dengan **machine learning**?

### Orthogonalization dalam Machine Learning

Pada **supervised learning**, biasanya diperlukan penyesuaian agar empat hal berikut terpenuhi:

1. **Kinerja di Training Set** - Algoritma harus tampil baik pada **training set** sebagai langkah awal.
2. **Kinerja di Dev Set** - Setelah itu, harapannya adalah kinerja yang baik pada **dev set**.
3. **Kinerja di Test Set** - Kinerja juga harus memadai pada **test set**.
4. **Kinerja di Dunia Nyata** - Pada akhirnya, performa di dunia nyata adalah tujuan utama.

> **Transkrip**:
> Dalam analogi TV, jika gambar terlalu lebar atau terlalu sempit, Anda ingin satu knob khusus untuk mengatur lebar tersebut. Dalam **machine learning**, jika algoritma tidak sesuai di **training set**, kita ingin memiliki knob khusus untuk mengatasi masalah ini, misalnya dengan menggunakan jaringan yang lebih besar atau algoritma **Adam**.

### Mengidentifikasi Knob untuk Meningkatkan Kinerja

Jika algoritma tidak berjalan baik pada **dev set**, ada **knob** lain yang bisa digunakan, misalnya **regularization** untuk meningkatkan kinerja pada dev set tanpa terlalu memengaruhi **training set**.

> **Transkrip**:
> Jika algoritma bekerja baik di **test set**, tetapi hasilnya tidak memuaskan pengguna, maka mungkin perlu meninjau ulang **dev set** atau **cost function**.

> **Penting**: Dengan menggunakan **orthogonalization**, kita dapat mengidentifikasi knob yang tepat untuk mengatasi permasalahan tertentu dalam sistem **machine learning**.



---

**Catatan**: **Early stopping** adalah teknik yang umum digunakan, tetapi bisa membingungkan karena mempengaruhi kinerja di **training set** dan **dev set** secara bersamaan, sehingga sulit untuk di-*tune* secara terpisah.
