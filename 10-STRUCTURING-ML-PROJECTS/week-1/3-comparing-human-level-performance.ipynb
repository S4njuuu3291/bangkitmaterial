# Perbandingan Sistem Pembelajaran Mesin dengan Kinerja Manusia

Dalam beberapa tahun terakhir, banyak tim pembelajaran mesin yang membicarakan tentang perbandingan sistem pembelajaran mesin dengan kinerja manusia. Mengapa hal ini terjadi? Saya pikir ada dua alasan utama. Yang pertama adalah karena kemajuan dalam **deep learning**, algoritma pembelajaran mesin secara tiba-tiba bekerja jauh lebih baik, sehingga menjadi jauh lebih memungkinkan di banyak area aplikasi bagi algoritma pembelajaran mesin untuk bersaing dengan kinerja manusia. Alasan kedua adalah ternyata alur kerja dalam merancang dan membangun sistem pembelajaran mesin menjadi jauh lebih efisien ketika Anda mencoba melakukan sesuatu yang juga bisa dilakukan oleh manusia. Jadi dalam pengaturan ini, menjadi wajar untuk membicarakan tentang perbandingan, atau mencoba meniru kinerja manusia. Mari kita lihat beberapa contoh untuk memahami ini.

## Progres dalam Pembelajaran Mesin

Saya telah melihat pada banyak tugas pembelajaran mesin bahwa seiring Anda bekerja pada suatu masalah selama waktu tertentu, yang sumbu x-nya adalah waktu, yang bisa berupa berbulan-bulan atau bahkan bertahun-tahun di mana tim atau komunitas riset bekerja pada masalah tersebut. **Kemajuan cenderung cukup cepat** saat mendekati kinerja manusia. Tetapi setelah beberapa saat, algoritma melewati kinerja manusia dan kemudian kemajuan serta akurasi sebenarnya melambat. Mungkin algoritma tersebut terus menjadi lebih baik, tetapi setelah melampaui kinerja manusia, meskipun bisa terus membaik, **kecepatan kenaikan akurasi seringkali melambat**. Harapannya adalah untuk mencapai tingkat performa optimal teoritis.

## Batasan Kesalahan Optimal Bayes

Seiring waktu, saat Anda terus melatih algoritma, mungkin dengan model yang lebih besar dan lebih banyak data, performa akan mendekati tetapi tidak pernah melampaui batas teoritis yang disebut **Bayes optimal error**.

**Bayes optimal error** adalah kesalahan terbaik yang bisa dicapai.

Ini adalah cara untuk setiap fungsi yang memetakan dari x ke y yang melampaui tingkat akurasi tertentu. Sebagai contoh, dalam pengenalan suara, jika x adalah klip audio, beberapa audio begitu berisik sehingga sangat sulit untuk mengetahui apa yang ada dalam transkripsi yang benar. Oleh karena itu, kesalahan sempurna mungkin tidak 100%. Atau untuk pengenalan kucing. Mungkin beberapa gambar begitu buram, sehingga tidak mungkin bagi siapa pun atau apapun untuk memberitahu apakah ada kucing di gambar tersebut. Jadi, tingkat akurasi yang sempurna mungkin tidak 100%. **Bayes optimal error**, atau kadang disebut **Bayes error** adalah fungsi teoretis terbaik untuk memetakan dari x ke y yang tidak pernah bisa dilampaui.



### Kinerja dan Pembatasan Bayes Optimal Error

Jadi, tidak mengherankan bahwa garis ungu ini, tidak peduli berapa lama Anda bekerja pada suatu masalah, Anda tidak bisa melampaui **Bayes optimal error**. Ternyata, kemajuan seringkali sangat cepat hingga Anda melampaui kinerja manusia.

Namun, kadang kemajuan melambat setelah melampaui kinerja manusia. Saya pikir ada dua alasan mengapa hal ini terjadi. 

### Mengapa Kemajuan Melambat Setelah Melampaui Kinerja Manusia?

1. **Kinerja Manusia Dekat dengan Bayes Optimal Error**
   Untuk banyak tugas, kinerja manusia sudah sangat dekat dengan **Bayes optimal error**. Manusia sangat baik dalam melihat gambar dan memberitahu apakah ada kucing di dalamnya atau mendengarkan audio dan menyalinnya. Jadi, setelah Anda melampaui kinerja manusia, mungkin tidak banyak ruang untuk perbaikan lebih lanjut.

2. **Keterbatasan Alat setelah Melampaui Kinerja Manusia**
   Selama kinerja algoritma Anda lebih buruk daripada manusia, ada alat tertentu yang bisa Anda gunakan untuk meningkatkan performa yang menjadi lebih sulit diterapkan setelah Anda melampaui kinerja manusia. Berikut penjelasannya:

   Untuk tugas yang sangat dikuasai manusia, seperti melihat gambar dan mengenali objek, mendengarkan audio, atau membaca bahasa, manusia cenderung sangat baik. Selama algoritma pembelajaran mesin Anda masih lebih buruk daripada manusia, Anda bisa mendapatkan **data berlabel** dari manusia. Artinya, Anda bisa meminta orang untuk memberi label pada contoh untuk Anda sehingga Anda bisa memiliki lebih banyak data untuk memberi makan algoritma pembelajaran Anda. Hal ini akan dibahas minggu depan dalam **analisis kesalahan manual**. Tetapi selama manusia masih melakukan lebih baik daripada algoritma mana pun, Anda bisa meminta orang untuk melihat contoh yang salah dikenali oleh algoritma Anda, dan mencoba mendapatkan wawasan mengapa orang bisa benar tetapi algoritma salah. Ini akan membantu meningkatkan kinerja algoritma Anda.

   Anda juga bisa mendapatkan **analisis bias dan variansi** yang lebih baik, yang akan dibahas sebentar lagi. Tetapi selama algoritma Anda masih lebih buruk daripada manusia, Anda memiliki taktik penting untuk meningkatkan algoritma Anda. Sedangkan setelah algoritma Anda melampaui kinerja manusia, taktik-taktik ini menjadi lebih sulit diterapkan.

### Perbandingan dengan Kinerja Manusia

Jadi, ini mungkin alasan lain mengapa perbandingan dengan kinerja manusia sangat berguna, terutama untuk tugas-tugas yang dikuasai manusia.



Inilah mengapa algoritma pembelajaran mesin cenderung sangat baik dalam mencoba meniru tugas yang bisa dilakukan manusia dan akhirnya bisa **mengejar** dan mungkin sedikit **melampaui kinerja manusia**. Secara khusus, meskipun Anda mengetahui apa itu bias dan variansi, ternyata mengetahui seberapa baik manusia dapat melaksanakan suatu tugas dapat membantu Anda memahami lebih baik seberapa banyak Anda harus mengurangi bias dan seberapa banyak Anda harus mengurangi variansi.

## Pembahasan Tentang Algoritma Pembelajaran dan Kinerja Manusia

Kita telah membahas bagaimana algoritma pembelajaran Anda seharusnya dapat bekerja dengan baik pada *training set*, tetapi kadang-kadang Anda tidak ingin algoritma tersebut bekerja terlalu baik. Mengetahui kinerja manusia bisa memberi tahu Anda seberapa baik, tetapi tidak terlalu baik, yang Anda inginkan agar algoritma Anda lakukan pada *training set*. Izinkan saya menunjukkan maksud saya.

### Contoh Klasifikasi Kucing

Kita telah banyak menggunakan **Cat classification**. Misalnya, katakanlah manusia memiliki tingkat akurasi hampir sempurna sehingga **human level error** adalah satu persen. Dalam kasus ini, jika algoritma pembelajaran Anda mencapai **training error** 8% dan **dev error** 10%, mungkin Anda ingin algoritma tersebut melakukan lebih baik pada *training set*. Jadi, fakta bahwa ada perbedaan besar antara bagaimana algoritma Anda melakukan pada *training set* dibandingkan dengan manusia menunjukkan bahwa algoritma Anda bahkan tidak cocok dengan *training set* dengan baik.

### Mengurangi Bias atau Varians?

Terkait dengan alat untuk mengurangi **bias** atau **variance**, dalam kasus ini saya akan mengatakan fokuslah pada pengurangan bias. Anda ingin melakukan hal-hal seperti melatih **neural network** yang lebih besar atau menjalankan pelatihan lebih lama, coba lakukan lebih baik pada *training set*.

### Kasus dengan **Human Level Error** yang Berbeda

Sekarang, mari kita lihat **training error** dan **dev error** yang sama dan bayangkan bahwa **human level performance** bukan 1%. Misalnya, dalam aplikasi atau data set yang berbeda, **human level error** ternyata 7,5%. Mungkin gambar dalam *data set* Anda sangat buram sehingga bahkan manusia pun tidak bisa mengatakan apakah ada kucing dalam gambar tersebut. Contoh ini mungkin sedikit dipaksakan karena manusia sebenarnya sangat baik dalam mengenali gambar dan memberitahukan apakah ada kucing di dalamnya atau tidak. Tetapi demi contoh ini, misalkan gambar dalam *data set* Anda sangat buram atau sangat rendah resolusinya sehingga bahkan manusia mendapatkan kesalahan 7,5%.

Dalam kasus ini, meskipun **training error** dan **dev error** Anda sama seperti contoh sebelumnya, Anda mungkin sebenarnya sudah melakukan dengan cukup baik pada *training set*. Algoritma Anda hanya sedikit lebih buruk daripada **human level performance**. 

### Fokus pada Pengurangan Varians

Pada contoh kedua ini, Anda mungkin ingin fokus pada pengurangan **variance** dalam algoritma pembelajaran Anda. Misalnya, Anda bisa mencoba **regularization** untuk mencoba mendekatkan **dev error** ke **training error**.

### Bias dan Varians dalam Pembahasan Sebelumnya

Dalam diskusi sebelumnya tentang **bias** dan **variance**, kita pada dasarnya mengasumsikan bahwa ada tugas di mana **Bayes error** hampir nol. Untuk menjelaskan apa yang baru saja terjadi, pada contoh **Cat classification**, anggap **human level error** sebagai perkiraan atau estimasi dari **Bayes error** atau **Bayes optimal error**. Untuk tugas **computer vision**, ini adalah perkiraan yang cukup masuk akal karena manusia sangat baik dalam **computer vision**, jadi apa pun yang bisa dilakukan manusia mungkin tidak terlalu jauh dari **Bayes error**.

### Mengurangi Bias atau Varians Berdasarkan Perkiraan **Bayes Error**

Yang mengejutkan di sini adalah bahwa tergantung pada **human level error** atau yang sebenarnya kira-kira **Bayes error** (atau yang kita anggap demikian), dengan **training error** dan **dev error** yang sama dalam kedua kasus ini, kita memutuskan untuk fokus pada pengurangan bias atau pengurangan varians. 

- Dalam contoh di sebelah kiri, **8% training error** cukup tinggi jika Anda berpikir bisa menurunkannya menjadi 1%, jadi taktik pengurangan bias bisa membantu Anda melakukan hal itu.
- Sedangkan dalam contoh di sebelah kanan, jika Anda berpikir bahwa **Bayes error** adalah 7,5%, maka Anda tahu bahwa tidak banyak ruang untuk menurunkan **training error** lebih jauh. Anda tidak ingin melakukannya lebih baik dari 7,5% karena itu hanya bisa dicapai dengan **overfitting** *training set*, dan sebaliknya, ada lebih banyak ruang untuk perbaikan dalam mengurangi selisih 2% ini dengan menggunakan teknik pengurangan **variance** seperti **regularization** atau mungkin mendapatkan lebih banyak data pelatihan.

### Pengertian **Avoidable Bias** dan **Variance**

Untuk memberikan beberapa istilah, ini bukan terminologi yang luas digunakan, tetapi saya merasa ini berguna sebagai cara berpikir. Saya akan menyebut perbedaan antara **Bayes error** atau perkiraan **Bayes error** dan **training error** sebagai **avoidable bias**. Jadi, yang Anda inginkan adalah terus meningkatkan kinerja pelatihan Anda hingga mencapai **Bayes error**, tetapi Anda tidak benar-benar ingin melakukannya lebih baik dari **Bayes error**. Anda tidak bisa melakukannya lebih baik dari **Bayes error** kecuali Anda **overfitting**.

Perbedaan antara **training error** dan **dev error** menunjukkan masalah **variance** dari algoritma Anda. **Avoidable bias** mengakui bahwa ada beberapa bias atau tingkat kesalahan minimum yang tidak dapat Anda turunkan, yang berarti jika **Bayes error** adalah 7,5%, Anda tidak ingin turun di bawah tingkat kesalahan itu.



### Perbedaan Pengurangan Bias dan Varians

- Dalam contoh di sebelah kiri, **7%** adalah ukuran dari **avoidable bias**, sedangkan 2% adalah ukuran dari seberapa besar **variance** yang Anda miliki. Jadi, pada contoh ini, ada lebih banyak potensi untuk fokus pada pengurangan **avoidable bias**.
- Sebaliknya, dalam contoh di sebelah kanan, **2%** lebih banyak ruang untuk pengurangan dibandingkan dengan **0,5%** yang mewakili **avoidable bias**.

### Fokus pada Pengurangan Bias atau Varians

Memahami **human level error**, memahami perkiraan **Bayes error**, benar-benar membuat Anda fokus pada taktik yang berbeda dalam situasi yang berbeda, apakah itu taktik penghindaran bias atau taktik penghindaran varians. Ada banyak nuansa lebih dalam bagaimana Anda memasukkan kinerja manusia dalam keputusan Anda untuk memilih fokus yang tepat.


# Pemahaman Tentang Human-Level Performance

Istilah **human-level performance** kadang digunakan secara santai dalam artikel penelitian. Namun, mari kita lihat bagaimana kita dapat mendefinisikannya dengan lebih tepat. Secara khusus, kita akan menggunakan definisi istilah **human-level performance** yang paling berguna untuk membantu Anda mendorong kemajuan dalam proyek machine learning Anda.

## Apa Itu Human-Level Error?

Ingat dari video terakhir kita bahwa salah satu penggunaan frasa **human-level error** adalah memberikan cara untuk mengestimasi **Bayes error**. Apa kesalahan terbaik yang mungkin dicapai oleh suatu fungsi, baik sekarang atau di masa depan? Jadi, dengan itu dalam pikiran, mari kita lihat contoh klasifikasi citra medis. Asumsikan bahwa manusia biasa yang tidak terlatih mencapai kesalahan 3% pada tugas ini. Dokter biasa, mungkin seorang radiologis, mencapai kesalahan 1%. Seorang dokter berpengalaman melakukan lebih baik lagi dengan kesalahan 0,7%. Dan sebuah tim dokter berpengalaman, yaitu jika Anda mengumpulkan tim dokter berpengalaman dan membiarkan mereka semua melihat gambar serta berdiskusi dan berdebat bersama, maka konsensus mereka mencapai kesalahan 0,5%. 

Pertanyaan yang ingin saya ajukan adalah, bagaimana Anda mendefinisikan **human-level error**? Apakah human-level error itu 3%, 1%, 0,7%, atau 0,5%? Silakan beri waktu untuk berpikir jika Anda ingin. Untuk menjawab pertanyaan tersebut, saya mendorong Anda untuk mengingat bahwa salah satu cara yang paling berguna untuk memandang kesalahan manusia adalah sebagai proksi atau estimasi dari **Bayes error**. Berikut adalah bagaimana saya akan mendefinisikan **human-level error**. Jika Anda ingin proksi atau estimasi untuk **Bayes error**, maka mengingat bahwa sebuah tim dokter berpengalaman yang berdiskusi dapat mencapai kesalahan 0,5%, kita tahu bahwa **Bayes error** lebih kecil atau sama dengan 0,5%. Jadi, karena beberapa sistem atau tim dokter dapat mencapai kesalahan 0,5%, maka berdasarkan definisi ini, kesalahan optimal pasti 0,5% atau lebih rendah. Kita tidak tahu seberapa lebih baik, mungkin ada tim yang lebih besar dengan dokter yang lebih berpengalaman yang bisa melakukannya lebih baik, jadi mungkin sedikit lebih baik dari 0,5%. Jadi, dalam hal ini saya akan menggunakan 0,5% sebagai estimasi **Bayes error**. Saya akan mendefinisikan **human-level performance** sebagai 0,5%, setidaknya jika Anda berharap menggunakan **human-level error** dalam analisis bias dan varians seperti yang kita lihat di video sebelumnya.

Untuk tujuan penerbitan artikel penelitian atau untuk tujuan penerapan sistem, mungkin ada definisi berbeda untuk **human-level error** yang bisa Anda gunakan, yaitu selama Anda melampaui kinerja dokter biasa. Itu bisa menjadi hasil yang sangat berguna jika tercapai, dan mungkin melampaui kinerja satu radiologis, seorang dokter, bisa berarti bahwa sistem cukup baik untuk diterapkan dalam beberapa konteks.

Tetapi jika tujuan Anda adalah proksi untuk **Bayes error**, maka definisi ini yang sesuai.

## Analisis Error dalam Kasus Klasifikasi Citra Medis

Misalnya, untuk contoh diagnosis citra medis, jika **training error** Anda adalah 5% dan **dev error** Anda adalah 6%. Dalam contoh dari slide sebelumnya, kinerja **human-level** kita, dan saya akan menganggap ini sebagai proksi untuk **Bayes error**.

Tergantung pada apakah Anda mendefinisikannya sebagai kinerja dokter biasa, dokter berpengalaman, atau tim dokter, Anda akan memiliki nilai 1%, 0,7%, atau 0,5% untuk ini. Ingat juga definisi kita dari video sebelumnya, bahwa selisih antara **Bayes error** atau estimasi **Bayes error** dan **training error** adalah ukuran dari **avoidable bias**. Sedangkan selisih ini merupakan ukuran atau estimasi masalah **variance** dalam algoritma pembelajaran Anda.



### Contoh 1: Training Error 5%, Dev Error 6%

Dalam contoh pertama ini, apapun definisi yang Anda pilih, ukuran **avoidable bias** akan sekitar 4%. Ini akan berada di antara 4%, jika Anda menggunakan 0,5%, sedangkan ini 1%. Jadi, dalam contoh ini, saya katakan, tidak masalah definisi **human-level error** yang Anda gunakan, apakah menggunakan kesalahan dokter biasa atau dokter berpengalaman atau tim dokter berpengalaman. Apapun ini 4% atau 4,5%, jelas ini lebih besar daripada masalah **variance**. Jadi, dalam hal ini, Anda harus fokus pada teknik pengurangan bias seperti melatih jaringan yang lebih besar.

### Contoh 2: Training Error 1%, Dev Error 5%

Sekarang, mari kita lihat contoh kedua. Misalkan **training error** Anda adalah 1% dan **dev error** Anda adalah 5%. Sekali lagi, tidak masalah apakah Anda menggunakan definisi **human-level performance** 1%, 0,7%, atau 0,5%. Apapun definisinya, ukuran **avoidable bias** akan sekitar 0% hingga 0,5%. Namun, selisih 4% antara **dev error** dan **training error** jauh lebih besar daripada **avoidable bias**. Jadi, ini menunjukkan bahwa Anda harus fokus pada pengurangan **variance** seperti regularisasi atau mendapatkan lebih banyak data latih.

### Contoh 3: Training Error 0,7%, Dev Error 0,8%

Namun, masalah sebenarnya muncul jika **training error** Anda 0,7% dan **dev error** Anda 0,8%. Dalam hal ini, sangat penting untuk menggunakan estimasi **Bayes error** sebagai 0,5%.

Karena dalam kasus ini, ukuran **avoidable bias** Anda adalah 0,2%, yang dua kali lebih besar dari ukuran **variance** Anda, yaitu 0,1%. Ini menunjukkan bahwa mungkin baik bias dan varians adalah masalah, tetapi **avoidable bias** sedikit lebih besar masalahnya.

## Mengapa Ini Penting?

Semoga ini memberikan gambaran mengapa membuat kemajuan dalam masalah machine learning menjadi lebih sulit saat Anda mendekati **human-level performance**. Dalam contoh ini, setelah Anda mendekati 0,7% error, kecuali Anda sangat hati-hati dalam memperkirakan **Bayes error**, Anda mungkin tidak tahu seberapa jauh Anda dari **Bayes error**. ## Kesimpulan

Untuk merangkum apa yang telah kita bahas, jika Anda mencoba memahami **bias** dan **variance** dengan estimasi **human-level error** untuk tugas yang bisa dilakukan manusia dengan sangat baik, Anda bisa menggunakan **human-level error** sebagai proksi atau perkiraan untuk **Bayes error**.



Selisih antara estimasi **Bayes error** memberi tahu Anda seberapa banyak **avoidable bias** adalah masalah, seberapa besar **avoidable bias** yang ada. # Menyusuri Kinerja Mesin Lebih Baik dari Manusia

Banyak tim sering merasa teruja untuk mengalahkan kinerja manusia pada tugas klasifikasi rekreasi tertentu. Mari kita bahas beberapa hal yang bisa Anda temui jika mencoba mencapainya sendiri. 

## Tantangan dalam Mencapai Kinerja Lebih Baik dari Manusia

Kita sebelumnya sudah membahas bagaimana kemajuan machine learning menjadi lebih sulit saat kita mendekati atau bahkan melampaui kinerja manusia. Mari kita diskusikan lagi satu contoh mengapa hal ini terjadi.

Misalkan Anda memiliki sebuah masalah di mana tim manusia yang berdiskusi dan berdebat mencapai **kesalahan 0.5%**, seorang manusia mencapai **kesalahan 1%**, dan algoritma Anda memiliki **kesalahan training 0.6%** dan **kesalahan dev 0.8%**. Dalam kasus ini, apa yang disebut sebagai *avoidable bias*?

### Jawaban untuk Bias yang Dapat Dihindari

Ini relatif lebih mudah untuk dijawab: **0.5%** adalah perkiraan dari **Bayes' error**, jadi *avoidable bias* Anda adalah, Anda tidak akan menggunakan angka **1%** sebagai referensi, melainkan menggunakan selisih ini. Jadi mungkin Anda memperkirakan *avoidable bias* Anda adalah setidaknya **0.1%** dan variansinya **0.2%**. Mungkin lebih banyak yang harus dilakukan untuk mengurangi variansinya dibandingkan dengan *avoidable bias*.

### Contoh yang Lebih Sulit

Sekarang, mari kita ambil contoh yang lebih sulit: Misalkan tim manusia dan kinerja manusia tunggal tetap seperti sebelumnya, tetapi algoritma Anda mendapatkan **kesalahan training 0.3%** dan **kesalahan dev 0.4%**. Sekarang, apa *avoidable bias* di sini?

Ini sekarang jauh lebih sulit untuk dijawab. Apakah fakta bahwa kesalahan training Anda 0.3% berarti Anda telah overfitting sebesar 0.2%? Atau apakah **Bayes' error** sebenarnya 0.1%, atau mungkin 0.2%, atau mungkin 0.3%? Anda tidak benar-benar tahu, tetapi berdasarkan informasi yang diberikan dalam contoh ini, Anda sebenarnya tidak memiliki cukup informasi untuk menentukan apakah Anda harus fokus pada pengurangan bias atau varians dalam algoritma Anda.

# Panduan Meningkatkan Performa Algoritma Pembelajaran

Pada video-video sebelumnya, kita telah membahas berbagai konsep seperti **orthogonalization**, cara menyiapkan **dev** dan **test** sets, **human-level performance** sebagai proksi untuk **Bayes error**, serta cara untuk memperkirakan **avoidable bias** dan **variance**. Sekarang, mari kita rangkum semuanya ke dalam satu set pedoman tentang cara meningkatkan performa algoritma pembelajaran Anda.

## Dua Hal Utama yang Perlu Diharapkan dari Algoritma Pembelajaran Supervised

Untuk membuat algoritma pembelajaran supervised bekerja dengan baik, ada dua hal yang secara fundamental perlu diharapkan atau diasumsikan dapat dilakukan oleh algoritma:

1. **Mengatasi Training Set dengan Baik**  
   Anda harus bisa menyesuaikan model dengan baik pada **training set**, yang berarti mencapai **avoidable bias** yang rendah.

2. **Generalize ke Dev atau Test Set**  
   Algoritma diharapkan bisa menggeneralisasi dengan baik dari **training set** ke **dev set** atau **test set**. Ini berarti **variance** tidak terlalu buruk.



## Mengatasi Masalah Avoidable Bias dan Variance

Dalam semangat **orthogonalization**, ada beberapa teknik yang bisa Anda coba untuk mengatasi masalah **avoidable bias** dan **variance**:

- **Untuk Mengatasi Avoidable Bias**:  
   Anda bisa mencoba beberapa teknik berikut:
   - Latih model yang lebih besar (train a bigger model).
   - Latih lebih lama (train longer).
   - Gunakan algoritma optimisasi yang lebih baik, seperti **Adam**, **RMSprop**, atau **ADS momentum**.
   - Cari arsitektur neural network yang lebih baik atau set hyperparameter yang lebih baik, misalnya mengubah fungsi aktivasi, jumlah lapisan, atau unit tersembunyi.

- **Untuk Mengatasi Variance**:  
   Jika **variance** menjadi masalah, beberapa teknik yang bisa dicoba antara lain:
   - Dapatkan lebih banyak data untuk pelatihan (**get more data**) karena lebih banyak data dapat membantu generalisasi yang lebih baik.
   - Terapkan **regularization**, seperti **L2 regularization**, **dropout**, atau **data augmentation**.
   - Cari arsitektur neural network atau set hyperparameter yang lebih cocok untuk masalah Anda.

## Merangkum Proses untuk Meningkatkan Performa



Untuk meningkatkan performa sistem pembelajaran mesin Anda, saya sarankan untuk:
1. Memeriksa selisih antara **training error** dan proksi **Bayes error** sebagai gambaran dari **avoidable bias**.
2. Mengamati selisih antara **dev error** dan **training error** sebagai perkiraan masalah **variance**.

Dengan demikian, Anda bisa mendapatkan gambaran lebih jelas tentang seberapa banyak usaha yang perlu dilakukan untuk memperbaiki performa dari **training set** ke **dev set** yang tidak dilatih secara eksplisit.
