# Etika dalam Pembelajaran Mesin

Album pembelajaran mesin saat ini memengaruhi miliaran orang. Anda pernah mendengar saya menyebutkan etika di video lain sebelumnya. Dan saya harap jika Anda membangun sistem pembelajaran mesin yang memengaruhi orang-orang, Anda berpikir untuk memastikan bahwa sistem Anda cukup adil, cukup bebas dari bias, dan bahwa Anda mengambil pendekatan etis untuk aplikasi Anda.

## Masalah Keadilan dan Bias

Mari kita lihat beberapa masalah yang berkaitan dengan keadilan, bias, dan etika. Sayangnya, dalam sejarah pembelajaran mesin, beberapa sistem yang telah dipublikasikan secara luas menunjukkan tingkat bias yang sama sekali tidak dapat diterima. Misalnya, ada dua sistem perekrutan yang terbukti mendiskriminasi perempuan. Perusahaan yang membangun sistem tersebut berhenti menggunakannya, tetapi harapan orang adalah bahwa sistem itu tidak pernah diluncurkan sejak awal.



Ada juga contoh sistem pengenalan wajah yang terdokumentasi dengan baik, yang mencocokkan individu berkulit gelap dengan tembakan cangkir kriminal lebih sering daripada individu berkulit lebih terang. Ini jelas tidak dapat diterima dan menjadi tanggung jawab komunitas untuk tidak membangun dan menerapkan sistem dengan masalah seperti ini.

### Contoh Kasus Negatif

1. **Persetujuan Pinjaman**  
   Sistem yang memberikan persetujuan pinjaman bank dengan cara yang bias dan diskriminatif terhadap subkelompok tertentu.

2. **Penguatan Stereotip Negatif**  
   Menggunakan algoritme yang dapat memperkuat stereotip negatif. Misalnya, jika seorang putri mencari profesi tertentu secara online dan tidak melihat siapa pun yang mirip dengannya, hal itu dapat memengaruhi pandangannya terhadap profesi tersebut.

3. **Kejahatan Online**  
   Terdapat kasus penggunaan algoritma pembelajaran mesin yang merugikan, seperti pembuatan konten palsu untuk tujuan komersial atau politik.

4. **Spam dan Penipuan**  
   Di industri email, terjadi pertempuran antara spammer dan komunitas anti-spam. Misalnya, di industri keuangan, penipu menggunakan pembelajaran mesin untuk melakukan penipuan.



## Pendekatan Etis dalam Pembelajaran Mesin

Demi kebaikan, hindari membangun sistem pembelajaran mesin yang berdampak negatif bagi masyarakat. Jika Anda diminta untuk mengerjakan aplikasi yang Anda anggap tidak etis, saya mendorong Anda untuk mempertimbangkan nilai-nilai etika Anda. Ada kalanya saya melihat proyek yang tampaknya sehat secara finansial, tetapi saya telah memilih untuk tidak terlibat dalam proyek tersebut karena alasan etis.

### Pentingnya Etika

Etika adalah subjek yang sangat rumit dan kaya yang telah dipelajari umat manusia selama ribuan tahun. Ketika AI semakin luas, saya membaca banyak buku tentang filsafat dan etika, berharap menemukan panduan yang jelas tentang cara berperilaku etis. Namun, saya menyadari bahwa tidak ada daftar periksa sederhana yang dapat memberikan panduan konkret tentang hal ini.

## Saran untuk Praktik yang Lebih Adil

Berikut adalah beberapa saran untuk membuat pekerjaan Anda lebih adil, kurang bias, dan lebih etis:

1. **Tim yang Beragam**  
   Kumpulkan tim yang beragam untuk brainstorming kemungkinan bahaya. Keragaman dalam berbagai dimensi, termasuk gender, etnis, dan budaya, dapat meningkatkan kemampuan tim dalam mengidentifikasi masalah sebelum sistem diluncurkan.

2. **Standar Industri**  
   Lakukan pencarian literatur tentang standar atau pedoman dalam industri Anda. Misalnya, dalam industri keuangan, sudah mulai ada standar untuk menentukan keadilan dan kebebasan dari bias dalam sistem.

3. **Audit Sistem**  
   Setelah mengidentifikasi masalah, audit sistem untuk mengukur kinerja terhadap dimensi yang diidentifikasi. Pastikan untuk mengidentifikasi dan memperbaiki setiap masalah sebelum penerapan.

4. **Rencana Mitigasi**  
   Kembangkan rencana mitigasi yang jelas. Rencana mitigasi sederhana dapat berupa kembali ke sistem sebelumnya yang sudah terbukti adil. Terus monitor setelah penerapan untuk mendeteksi bahaya dan bertindak cepat jika diperlukan.



## Kesimpulan

Saya telah mengerjakan banyak sistem pembelajaran mesin dan ingin menekankan pentingnya memperhatikan masalah etika, keadilan, dan bias. Ini bukan sesuatu yang bisa dianggap remeh. Proyek dengan implikasi etika yang lebih serius harus diperlakukan dengan lebih hati-hati. Jika Anda membangun sistem untuk memutuskan pinjaman, misalnya, implikasi etisnya jauh lebih besar dibandingkan dengan sistem yang hanya memutuskan berapa lama memanggang biji kopi.

Saya berharap kita semua dapat terus memperdebatkan masalah ini dengan baik, menemukan dan memperbaiki masalah sebelum menyebabkan kerusakan. Sistem yang kita bangun dapat memengaruhi banyak orang, jadi sangat penting untuk memikirkan etika dalam setiap langkah pengembangan sistem pembelajaran mesin.
