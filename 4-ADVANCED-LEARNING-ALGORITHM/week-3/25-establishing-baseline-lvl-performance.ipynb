# Penjelasan Tentang Evaluasi Bias dan Variance pada Algoritma Pembelajaran

Mari kita lihat beberapa angka konkret untuk J-train dan JCV, serta bagaimana Anda dapat menilai apakah sebuah algoritma pembelajaran memiliki bias tinggi atau varian tinggi. Dalam contoh ini, saya akan menggunakan pengenalan suara sebagai contoh berkelanjutan, yang pernah saya kerjakan beberapa kali.

## Contoh Pengenalan Suara

Banyak pengguna yang melakukan pencarian web melalui ponsel lebih memilih menggunakan pengenalan suara daripada mengetik di keyboard kecil pada ponsel. Misalnya, input suara pengguna mungkin seperti ini: "Cuaca hari ini bagaimana?" atau "Toko kopi di dekat sini". Algoritma pengenalan suara bertugas mengubah input audio tersebut menjadi teks.

Jika Anda melatih sistem pengenalan suara dan mengukur kesalahan pelatihan (training error), katakanlah Anda mendapatkan kesalahan pelatihan sebesar **10.8%**. Artinya, sistem berhasil mengenali suara dengan benar pada **89.2%** dari data pelatihan, dan membuat kesalahan pada **10.8%** sisanya.

Sekarang, jika Anda juga mengukur kinerja algoritma pada set validasi silang (cross-validation set), Anda mungkin mendapatkan kesalahan validasi sebesar **14.8%**. Pada pandangan pertama, sepertinya kesalahan pelatihan dan validasi cukup tinggi, dengan kesalahan pelatihan mencapai **10.8%**. Hal ini mungkin membuat Anda berpikir bahwa algoritma memiliki **bias yang tinggi**, karena tidak berkinerja baik bahkan pada data pelatihan.

Namun, ada faktor lain yang perlu dipertimbangkan, yaitu kinerja manusia. Pada aplikasi pengenalan suara, penting untuk mengetahui seberapa baik manusia mampu mengenali suara dengan akurat. Misalnya, katakanlah Anda menemukan bahwa manusia memiliki kesalahan sebesar **10.6%** dalam mentranskripsikan klip audio yang sama.



### Pengaruh Kesalahan Manusia



Mengapa kesalahan manusia begitu tinggi? Ternyata, banyak klip audio yang berisi kebisingan, sehingga bahkan manusia tidak dapat mentranskripsikan audio tersebut dengan sempurna. Misalnya, klip seperti ini: "Saya akan ke [inaudible]." Karena adanya kebisingan, tidak ada yang dapat memahami sepenuhnya apa yang diucapkan.

Jika manusia memiliki kesalahan **10.6%**, maka tidak realistis untuk mengharapkan algoritma pembelajaran melakukan jauh lebih baik. Jadi, ketika kita membandingkan kesalahan pelatihan sebesar **10.8%** dengan kesalahan manusia **10.6%**, kita bisa menyimpulkan bahwa algoritma tersebut sebenarnya sudah cukup baik dalam hal **bias**.

Namun, ada perbedaan yang signifikan antara kesalahan validasi **14.8%** dan kesalahan pelatihan **10.8%**, yang menunjukkan bahwa algoritma ini lebih memiliki **masalah varian** daripada bias.

## Menetapkan Baseline Kinerja


Untuk menilai apakah kesalahan pelatihan tinggi, penting untuk menetapkan **baseline** (acuan kinerja). Baseline ini bisa berupa performa manusia, yang dalam contoh ini adalah **10.6%**. Selain itu, baseline juga bisa berupa algoritma kompetitor atau versi sebelumnya dari algoritma yang sedang diuji. Dengan baseline ini, kita dapat lebih akurat menilai apakah algoritma memiliki bias atau varian yang tinggi.

Jika kesalahan pelatihan lebih tinggi dari baseline, maka kita bisa menyimpulkan bahwa algoritma memiliki **bias yang tinggi**. Namun, jika perbedaan antara kesalahan pelatihan dan validasi besar, maka itu menunjukkan adanya **varian yang tinggi**.

## Contoh Lain



Jika kesalahan pelatihan lebih tinggi dibandingkan baseline (misalnya, **14.8%** kesalahan pelatihan dibandingkan **10.6%** kesalahan manusia), dan kesalahan validasi hanya sedikit lebih tinggi dari kesalahan pelatihan, maka itu menunjukkan algoritma memiliki **bias yang tinggi**.

Sebaliknya, jika kesalahan pelatihan dekat dengan baseline tetapi kesalahan validasi jauh lebih besar (seperti contoh pertama), maka itu menunjukkan **masalah varian**.

## Masalah Bias dan Varian yang Tinggi

Ada juga kemungkinan algoritma memiliki **bias dan varian yang tinggi**. Misalnya, jika kesalahan pelatihan jauh di atas baseline, dan kesalahan validasi juga jauh lebih tinggi dari kesalahan pelatihan, maka algoritma tersebut memiliki kedua masalah.

## Simpulan

Dengan melihat perbedaan antara kesalahan pelatihan, baseline, dan kesalahan validasi, Anda dapat menilai apakah algoritma memiliki **bias** atau **varian** yang tinggi. Pada beberapa aplikasi, terutama di mana data mungkin berisik, baseline kinerja manusia sering kali merupakan acuan yang baik. Jika algoritma Anda berkinerja buruk dibandingkan baseline ini, maka Anda mungkin memiliki masalah bias, dan jika kesalahan validasi jauh lebih besar daripada kesalahan pelatihan, Anda mungkin memiliki masalah varian.

### Learning Curve (Kurva Pembelajaran)

Untuk lebih memperdalam pemahaman kita tentang kinerja algoritma pembelajaran, konsep lain yang berguna untuk dipertimbangkan adalah **learning curve**. Ini akan kita bahas lebih lanjut pada video atau penjelasan berikutnya.
