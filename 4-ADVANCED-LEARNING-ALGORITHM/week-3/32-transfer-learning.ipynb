# Transfer Learning dalam Pembelajaran Mesin

Untuk aplikasi di mana data yang dimiliki tidak banyak, **Transfer Learning** adalah teknik yang sangat baik yang memungkinkan Anda menggunakan data dari tugas lain untuk membantu aplikasi Anda. Teknik ini sangat sering saya gunakan. Mari kita lihat bagaimana **Transfer Learning** bekerja.

## Bagaimana Transfer Learning Bekerja?

Misalkan Anda ingin mengenali angka yang ditulis tangan dari nol hingga sembilan, tetapi Anda tidak memiliki banyak data berlabel dari angka-angka tersebut. Berikut adalah apa yang bisa Anda lakukan:

1. Katakanlah Anda menemukan dataset yang sangat besar berisi satu juta gambar kucing, anjing, mobil, orang, dan sebagainya, dengan seribu kelas.
2. Anda bisa memulai dengan melatih jaringan neural (neural network) pada dataset besar ini dengan seribu kelas yang berbeda, dan mengajari algoritma untuk mengenali salah satu dari 1.000 kelas tersebut.

Pada proses ini, Anda akan belajar parameter untuk lapisan pertama jaringan neural: `W^1`, `b^1`, untuk lapisan kedua `W^2`, `b^2`, dan seterusnya hingga lapisan terakhir `W^5`, `b^5`.

### Menerapkan Transfer Learning

Untuk menerapkan **Transfer Learning**, yang harus Anda lakukan adalah:

- Membuat salinan jaringan neural ini dan mempertahankan parameter `W^1`, `b^1`, `W^2`, `b^2`, `W^3`, `b^3`, dan `W^4`, `b^4`.
- Hapus lapisan output terakhir dan ganti dengan lapisan output yang lebih kecil dengan hanya 10 unit output (untuk angka nol hingga sembilan).

Parameter `W^5`, `b^5` tidak dapat disalin karena dimensi lapisan ini berubah, sehingga Anda perlu membuat parameter baru `W^5`, `b^5` yang harus dilatih dari awal.

### Dua Opsi Pelatihan

Ada dua opsi untuk melatih parameter jaringan neural ini:

1. **Opsi 1:** Hanya melatih parameter lapisan output. Anda memegang parameter `W^1`, `b^1` hingga `W^4`, `b^4` dan tidak mengubahnya, lalu hanya memperbarui `W^5`, `b^5`.
   
2. **Opsi 2:** Melatih semua parameter jaringan neural, termasuk `W^1`, `b^1` hingga `W^5`, `b^5`, namun lapisan pertama hingga keempat diinisialisasi dengan nilai yang sudah dilatih sebelumnya.



Jika dataset Anda kecil, **Opsi 1** mungkin lebih baik, tetapi jika dataset Anda sedikit lebih besar, **Opsi 2** mungkin lebih efektif.

## Mengapa Transfer Learning Efektif?

Teknik ini disebut **Transfer Learning** karena idenya adalah dengan belajar mengenali kucing, anjing, mobil, dan orang, jaringan neural dapat belajar mendeteksi fitur-fitur dasar seperti tepi gambar, sudut, dan bentuk sederhana. Parameter ini kemudian dapat digunakan sebagai titik awal yang baik untuk mengenali angka yang ditulis tangan.



### Pembatasan Transfer Learning

Namun, ada satu pembatasan. Jenis data masukan (input) `x` harus sama untuk langkah pre-training dan fine-tuning. Misalnya, jika tugas akhir Anda adalah tugas penglihatan komputer, maka pre-training juga harus dilakukan dengan data gambar. Begitu pula jika Anda ingin membangun sistem pengenalan suara, maka pre-training harus menggunakan data audio.

## Langkah-langkah Transfer Learning

Berikut adalah dua langkah utama **Transfer Learning**:

1. **Unduh jaringan neural yang telah di-pre-trained** pada dataset besar dengan tipe input yang sama dengan aplikasi Anda (gambar, audio, teks, dll).
2. **Lanjutkan pelatihan** dengan dataset Anda sendiri, menggunakan opsi fine-tuning untuk menyesuaikan parameter jaringan neural agar sesuai dengan tugas spesifik Anda.



Dengan **Transfer Learning**, seringkali Anda dapat mencapai hasil yang cukup baik meskipun dataset yang Anda miliki kecil.

## Kesimpulan

**Transfer Learning** memungkinkan kita untuk memanfaatkan model yang sudah dilatih pada dataset besar, lalu melanjutkan pelatihan pada dataset yang lebih kecil. Ini adalah salah satu cara di mana komunitas pembelajaran mesin saling berbagi kerja dan hasil, memungkinkan kita semua untuk mencapai hasil yang lebih baik. Teknik ini telah diterapkan pada berbagai model terkenal seperti GPT-3, BERT, dan model pre-trained lainnya.
