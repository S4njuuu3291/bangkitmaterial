# Proses Pengembangan Sistem Pembelajaran Mesin

Dalam beberapa video berikutnya, saya ingin berbagi dengan Anda tentang bagaimana melalui proses pengembangan sistem pembelajaran mesin. Dengan harapan, ketika Anda melakukan hal tersebut sendiri, Anda bisa berada dalam posisi untuk membuat keputusan yang baik di berbagai tahapan proses pengembangan pembelajaran mesin.

## Siklus Iteratif Pengembangan Pembelajaran Mesin

Mari kita lihat terlebih dahulu siklus iteratif dari pengembangan pembelajaran mesin. Beginilah umumnya proses mengembangkan model pembelajaran mesin terasa:

1. **Menentukan Arsitektur Sistem**: 
   - Ini berarti memilih model pembelajaran mesin, serta memutuskan data apa yang akan digunakan, memilih hyperparameter, dan sebagainya.

2. **Mengimplementasikan dan Melatih Model**: 
   - Setelah keputusan tersebut diambil, Anda akan mengimplementasikan dan melatih model tersebut. Seperti yang telah saya sebutkan sebelumnya, ketika Anda melatih model untuk pertama kalinya, hampir tidak pernah bekerja sebaik yang Anda inginkan.

3. **Melakukan Diagnostik**: 
   - Langkah berikutnya yang saya rekomendasikan adalah melihat beberapa diagnostik, seperti melihat bias dan varians algoritma Anda serta analisis kesalahan, yang akan kita bahas di video berikutnya.

4. **Membuat Keputusan Berdasarkan Diagnostik**: 
   - Berdasarkan wawasan dari diagnostik, Anda bisa membuat keputusan, seperti apakah ingin memperbesar jaringan neural Anda, mengubah parameter regularisasi Lambda, menambah lebih banyak data, menambah fitur, atau bahkan mengurangi fitur.

5. **Iterasi Hingga Mendapatkan Performa yang Diinginkan**:
   - Anda akan terus berulang melalui siklus ini dengan pilihan arsitektur baru Anda, sering kali butuh beberapa iterasi hingga mencapai performa yang diinginkan.



## Contoh Pengembangan Klasifikasi Spam Email

Sebagai contoh, mari kita lihat bagaimana membangun sistem klasifikasi spam email.

Banyak dari kita yang sangat benci dengan spam email, dan ini adalah masalah yang pernah saya kerjakan beberapa tahun lalu. Saya juga pernah terlibat dalam mendirikan konferensi anti-spam beberapa tahun yang lalu.

- **Contoh Email Spam**: Gambar di sebelah kiri menunjukkan contoh email yang sangat spammy seperti "Deal of the Week, Buy Now, Rolex Watches." Terkadang, para spammer sengaja salah mengeja kata-kata seperti "watches," "medicine," atau "mortgages" untuk mencoba mengecoh sistem pengenal spam.

- **Contoh Email Biasa**: Di sisi kanan, adalah email yang pernah saya terima dari adik saya, Alfred, tentang ajakan berkumpul saat Natal. Jadi, bagaimana kita bisa membangun sebuah classifier untuk mengenali spam versus non-spam?



### Membuat Algoritma Pembelajaran Supervised

Salah satu caranya adalah dengan melatih algoritma pembelajaran terawasi (supervised learning), di mana fitur input (x) adalah fitur-fitur dari sebuah email, dan label output (y) adalah 1 atau 0, tergantung pada apakah itu spam atau bukan spam.

### Konstruksi Fitur Email

Ini adalah contoh **klasifikasi teks**, karena kita mengambil dokumen teks (email) dan mencoba mengklasifikasikannya sebagai spam atau bukan spam. Salah satu cara untuk membangun fitur dari email adalah dengan memilih 10.000 kata teratas dalam bahasa Inggris (atau dari kamus lainnya), dan menggunakan kata-kata tersebut sebagai fitur x₁, x₂ hingga x₁₀₀₀₀.

Sebagai contoh:

- Pada email di sebelah kanan, jika daftar kata kita adalah "a," "Andrew," "buy," "deal," "discount," dan seterusnya, maka kita bisa menetapkan fitur-fitur tersebut sebagai 0 atau 1 tergantung pada apakah kata tersebut muncul dalam email.
    - Kata "a" tidak muncul.
    - Kata "Andrew" muncul.
    - Kata "buy" muncul.
    - Kata "deal" muncul.
    - Kata "discount" tidak muncul.
    


Dengan cara ini, kita bisa membangun 10.000 fitur untuk email tersebut. Ada banyak cara untuk membangun vektor fitur seperti ini. Cara lain mungkin dengan menghitung berapa kali sebuah kata muncul dalam email.

### Melatih Model Klasifikasi

Setelah membangun fitur, Anda bisa melatih algoritma klasifikasi seperti **logistic regression** atau **neural network** untuk memprediksi y berdasarkan fitur x ini. Namun, setelah Anda melatih model awal, jika hasilnya tidak sesuai harapan, Anda akan memiliki banyak ide untuk memperbaiki kinerja algoritma pembelajaran.

## Ide Perbaikan Algoritma

Berikut adalah beberapa contoh ide yang mungkin muncul:

1. **Mengumpulkan Lebih Banyak Data**: 
   - Selalu menggoda untuk mencoba mengumpulkan lebih banyak data. Misalnya, beberapa teman saya pernah bekerja pada proyek **honeypot** berskala besar. Proyek ini membuat sejumlah besar alamat email palsu yang sengaja diberikan kepada spammer agar mereka mengirimkan spam. Ini menjadi cara mendapatkan banyak data spam.

2. **Mengembangkan Fitur Lebih Canggih**: 
   - Anda mungkin bekerja pada pengembangan fitur-fitur yang lebih canggih, seperti berdasarkan **routing email** (jalur email melintasi server yang berbeda) atau **header email**, yang bisa memberi petunjuk apakah email tersebut dikirim oleh spammer atau bukan.

3. **Mengolah Teks Email**: 
   - Anda mungkin juga bekerja pada cara mendeteksi salah eja atau kesalahan eja yang disengaja, seperti "watches," "medicine," dan "mortgages." Hal ini bisa membantu menentukan apakah email bersifat spammy.



## Pentingnya Diagnostik

Dengan semua ide di atas dan mungkin lebih banyak lagi, bagaimana Anda memutuskan mana yang lebih menjanjikan untuk dikerjakan? Memilih jalur yang lebih menjanjikan bisa mempercepat proyek Anda hingga 10 kali lipat dibandingkan jika Anda memilih jalur yang kurang menjanjikan.

Sebagai contoh:

- Jika algoritma Anda memiliki **bias yang tinggi** daripada varians yang tinggi, menghabiskan berbulan-bulan pada proyek honeypot mungkin bukan arah yang paling produktif.
- Tetapi, jika algoritma Anda memiliki **varians yang tinggi**, mengumpulkan lebih banyak data bisa sangat membantu.

## Analisis Kesalahan dan Iterasi Pengembangan

Selama proses iteratif pengembangan pembelajaran mesin, Anda mungkin memiliki banyak ide tentang bagaimana memodifikasi model atau data. Namun, menggunakan diagnostik yang tepat bisa memberi banyak panduan tentang pilihan model, data, atau bagian arsitektur lain yang paling menjanjikan untuk dicoba.
