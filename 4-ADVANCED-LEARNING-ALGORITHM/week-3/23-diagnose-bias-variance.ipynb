# **Workflow Pengembangan Sistem Machine Learning**

## 1. Tahapan Utama

Dalam pengembangan sistem machine learning, alurnya biasanya dimulai dengan ide dan pelatihan model. Namun, sering kali hasil awal dari model tidak sesuai harapan. Pada tahap ini, penting untuk menentukan langkah-langkah perbaikan yang diperlukan untuk meningkatkan kinerja model tersebut.

Kunci dalam membangun sistem machine learning adalah mengetahui cara menentukan apa yang harus dilakukan selanjutnya agar performa model dapat ditingkatkan. Berdasarkan pengalaman dalam berbagai aplikasi, salah satu cara efektif adalah dengan menganalisis **bias** dan **variance** dari algoritma yang digunakan.

### 1.1. Contoh Linear Regression

Sebagai contoh, misalkan kita memiliki dataset yang sederhana. Jika kita melakukan fit menggunakan garis lurus (linear regression), model tidak akan bekerja dengan baik. Dalam kasus ini, model memiliki **bias tinggi** (underfitting).

Jika kita melakukan fit menggunakan polinomial orde empat, hasilnya mungkin sangat sesuai dengan data pelatihan, namun justru gagal untuk data baru. Ini disebut **variance tinggi** (overfitting). Di sisi lain, model polinomial kuadrat mungkin akan lebih sesuai dengan dataset ini, menunjukkan keseimbangan yang tepat antara bias dan variance.

### 1.2. Penentuan Bias dan Variance

Pada masalah yang lebih kompleks, di mana kita memiliki lebih banyak fitur, sulit untuk memvisualisasikan fungsi dan melihat apakah model bekerja dengan baik atau tidak. Untuk mengatasi ini, pendekatan yang lebih sistematis adalah menganalisis kinerja model pada **training set** dan **cross-validation set**.

- **Bias Tinggi (Underfitting)**: Ketika model memiliki bias tinggi, performanya tidak baik baik pada data pelatihan (J_train tinggi) maupun pada data cross-validation (J_cv tinggi).
- **Variance Tinggi (Overfitting)**: Jika J_train rendah tetapi J_cv tinggi, model memiliki variance tinggi.

Dengan memeriksa perbedaan kinerja antara data pelatihan dan data validasi, kita bisa mengetahui apakah model memiliki bias atau variance yang tinggi.

---


## 2. Variasi Bias dan Variance Berdasarkan Tingkat Polinomial

### 2.1. Diagram Polinomial dan Error

Ketika kita meningkatkan orde polinomial dari model, error pada data pelatihan (J_train) cenderung menurun, karena model semakin rumit dan mampu "menghafal" data pelatihan. Namun, jika polinomial terlalu kompleks (misalnya orde ke-4 atau lebih), error pada data cross-validation (J_cv) justru meningkat, menunjukkan bahwa model mengalami **overfitting**.

Sebaliknya, jika polinomial terlalu sederhana (misalnya orde ke-1), baik J_train maupun J_cv akan tinggi, menunjukkan **underfitting**.



### 2.2. Kesimpulan

- **Bias Tinggi**: J_train tinggi, J_cv tinggi.
- **Variance Tinggi**: J_train rendah, J_cv tinggi.
- **Model Ideal**: J_train dan J_cv rendah, tanpa perbedaan signifikan antara keduanya.

---

## 3. Regularisasi

Regularisasi adalah teknik yang digunakan untuk mengatasi masalah bias dan variance. Penggunaan regularisasi dapat membantu mengurangi overfitting (variance tinggi) dengan mengontrol kompleksitas model. Pada bagian selanjutnya, kita akan mempelajari lebih dalam bagaimana regularisasi mempengaruhi bias dan variance.

---

## 4. Penutup

Ketika kita melatih model machine learning, penting untuk selalu mengevaluasi apakah model memiliki **bias tinggi** (tidak cukup belajar dari data pelatihan) atau **variance tinggi** (terlalu menyesuaikan dengan data pelatihan). Pemahaman ini akan memberikan panduan yang lebih baik untuk meningkatkan kinerja algoritma.
