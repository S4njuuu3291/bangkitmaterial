# Mengukur Kemurnian Kumpulan Contoh Menggunakan Entropi

Dalam video ini, kita akan membahas cara mengukur kemurnian dari suatu set contoh. Jika semua contoh tersebut adalah kucing dari satu kelas, maka itu sangat murni. Jika semua contoh bukan kucing, itu juga sangat murni. Namun, jika berada di antara kedua kondisi tersebut, bagaimana kita dapat mengkuantifikasi seberapa murni set contoh tersebut? Mari kita lihat definisi **entropi**, yang merupakan ukuran dari ketidakmurnian suatu set data.

## Definisi Entropi

Diberikan sebuah set contoh dengan enam data, kita memiliki tiga kucing dan tiga anjing. Mari kita definisikan $ p_1 $ sebagai fraksi contoh yang merupakan kucing, yaitu fraksi contoh dengan label satu. Ini yang dimaksud dengan subskrip satu. Dalam contoh ini, $ p_1 $ sama dengan $ \frac{3}{6} $.

Kita akan mengukur ketidakmurnian dari sekumpulan contoh menggunakan fungsi yang disebut **entropi**, yang dinyatakan dengan $ H(p_1) $. Fungsi entropi ini dapat digambarkan dalam bentuk kurva di mana sumbu horizontal adalah $ p_1 $ (fraksi kucing dalam sampel) dan sumbu vertikal adalah nilai dari entropi.

### Contoh dan Pengukuran Entropi

1. **Contoh 1: Setengah Kucing, Setengah Anjing**
   - Dalam contoh di mana $ p_1 = \frac{3}{6} $ atau 0.5, nilai entropi $ H(p_1) $ adalah 1. 
   - Kita lihat bahwa kurva ini mencapai puncaknya ketika set contoh memiliki komposisi 50-50, yang merupakan ketidakmurnian maksimum (entropi 1).

2. **Contoh 2: Lebih Banyak Kucing**
   - Jika kita memiliki set dengan lima kucing dan satu anjing, maka $ p_1 = \frac{5}{6} $ dan $ p_1 $ sekitar 0.83. 
   - Di sini, entropi $ H(p_1) $ adalah sekitar 0.65.

3. **Contoh 3: Semua Kucing**
   - Ketika semua enam contoh adalah kucing ($ p_1 = \frac{6}{6} $), entropi $ H(p_1) $ adalah 0.
   - Ini menunjukkan bahwa ketidakmurnian berkurang dari 1 (setengah-setengah) menjadi 0 (semua kucing).

4. **Contoh 4: Dua Kucing dan Empat Anjing**
   - Dalam contoh ini, $ p_1 = \frac{2}{6} $ atau $ \frac{1}{3} $, dan jika kita baca nilai entropi di 0.33, hasilnya sekitar 0.92. 
   - Ini menunjukkan bahwa set ini lebih tidak murni dibandingkan dengan contoh sebelumnya karena lebih dekat dengan komposisi 50-50.

5. **Contoh 5: Semua Anjing**
   - Ketika kita memiliki set yang terdiri dari enam anjing ($ p_1 = 0 $), entropi $ H(p_1) $ adalah 0, yang menunjukkan tidak ada ketidakmurnian.



## Persamaan untuk Fungsi Entropi

Mari kita lihat persamaan aktual untuk fungsi entropi $ H(p_1) $. Ingat bahwa $ p_1 $ adalah fraksi contoh yang merupakan kucing, jadi jika Anda memiliki sampel yang terdiri dari $ \frac{2}{3} $ kucing, maka sampel tersebut harus memiliki $ \frac{1}{3} $ contoh yang bukan kucing.

Kita mendefinisikan $ p_0 $ sebagai fraksi contoh yang bukan kucing, yang sama dengan $ 1 - p_1 $. Fungsi entropi kemudian didefinisikan sebagai:
$$
H(p_1) = -p_1 \log_2(p_1) - p_0 \log_2(p_0) 
$$

Atau bisa juga ditulis sebagai:
$$
H(p_1) = -p_1 \log_2(p_1) - (1 - p_1) \log_2(1 - p_1) 
$$



### Catatan Penting

Jika $ p_1 $ atau $ p_0 $ sama dengan 0, ekspresi seperti ini akan tampak seperti $ 0 \log(0) $, dan $ \log(0) $ secara teknis tidak terdefinisi, yaitu negatif tak terhingga. Namun, berdasarkan konvensi untuk tujuan penghitungan entropi, kita mengambil $ 0 \log(0) $ sebagai 0. 

Jika Anda berpikir bahwa definisi entropi ini sedikit mirip dengan definisi dari **logistic loss** yang kita pelajari di kursus sebelumnya, ada alasan matematis mengapa kedua rumus ini terlihat serupa. Namun, kita tidak akan membahasnya di kelas ini.

## Kesimpulan

Fungsi entropi adalah ukuran dari ketidakmurnian suatu set data. Fungsi ini dimulai dari nol, meningkat hingga satu, dan kemudian kembali turun ke nol sebagai fungsi dari fraksi contoh positif dalam sampel. Ada fungsi lain yang terlihat mirip, seperti **Gini criteria**, yang juga dapat digunakan untuk membangun pohon keputusan. Namun, untuk kesederhanaan, dalam video ini saya akan fokus pada menggunakan kriteria entropi yang biasanya cukup baik untuk sebagian besar aplikasi.
