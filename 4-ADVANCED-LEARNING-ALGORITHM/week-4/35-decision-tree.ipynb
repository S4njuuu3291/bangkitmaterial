## Pembelajaran tentang Pohon Keputusan

Dalam minggu ini, kita akan belajar tentang pohon keputusan dan bagaimana cara membuatnya bekerja untuk Anda. Mari kita menyelam lebih dalam.

### Contoh Klasifikasi Kucing



Untuk menjelaskan cara kerja pohon keputusan, saya akan menggunakan contoh klasifikasi kucing. Anda menjalankan pusat adopsi kucing dan diberikan beberapa fitur. Anda ingin melatih pengklasifikasi untuk dengan cepat memberi tahu Anda apakah seekor hewan itu kucing atau bukan. Berikut adalah **10 contoh pelatihan** yang dimiliki:

- Terkait dengan masing-masing dari 10 contoh ini, kita akan memiliki fitur mengenai:
  - **Bentuk telinga** hewan
  - **Bentuk wajah**
  - **Kehadiran kumis**

Kemudian, kita memiliki label kebenaran dasar yang ingin Anda prediksi, yaitu apakah hewan ini adalah kucing atau bukan.

#### Contoh Data

1. Contoh pertama: Telinga runcing, wajah bulat, kumis hadir, **itu adalah kucing**.
2. Contoh kedua: Telinga terkulai, bentuk wajah tidak bulat, kumis hadir, **itu adalah kucing**.
3. Dan seterusnya untuk contoh lainnya.



Dataset ini memiliki **lima kucing** dan **lima non-kucing** di dalamnya. Fitur input **X** adalah tiga kolom ini, dan output target yang ingin Anda prediksi, **Y**, apakah ini kucing atau bukan. Dalam contoh ini, fitur **X** mengambil nilai kategoris. 

### Tugas Klasifikasi Biner

Ini adalah tugas **klasifikasi biner** karena labelnya juga satu atau nol. Saat ini, masing-masing fitur **X_1**, **X_2**, dan **X_3** hanya mengambil dua nilai yang mungkin. Kami akan membahas fitur yang dapat mengambil lebih dari dua nilai yang mungkin, serta fitur bernilai berkelanjutan di akhir minggu ini.

## Apa itu Pohon Keputusan?

Berikut adalah contoh model yang mungkin Anda dapatkan setelah melatih algoritme pembelajaran pohon keputusan pada kumpulan data yang baru saja Anda lihat. Model yang dihasilkan oleh algoritma pembelajaran terlihat seperti pohon, dan gambar seperti ini disebut **pohon**.

### Struktur Pohon Keputusan

Jika tidak terlihat seperti pohon biologis yang Anda lihat di luar sana, tidak apa-apa, jangan khawatir tentang itu. Kami akan membahas contoh untuk memastikan bahwa definisi ilmu komputer tentang pohon ini juga masuk akal bagi Anda.

- Setiap oval atau persegi panjang ini disebut **simpul** di pohon. Cara kerja model ini adalah jika Anda memiliki contoh tes baru, ia memiliki kucing dengan bentuk telinga runcing, wajah bulat, dan kumis hadir.
  
- Model ini akan melihat contoh ini dan membuat keputusan klasifikasi dimulai dari simpul paling atas pohon, yang disebut **simpul akar** pohon. Kita akan melihat fitur yang tertulis di dalamnya, yaitu bentuk telinga.

### Proses Klasifikasi

Berdasarkan nilai bentuk telinga dari contoh ini, kita akan pergi ke kiri atau ke kanan. Jika bentuk telinganya runcing, kita akan turun cabang kiri pohon, dan berakhir di simpul oval ini. 

Kami kemudian melihat bentuk wajah dari contoh ini, yang ternyata bulat, jadi kami akan mengikuti panah ini ke bawah. Akhirnya, algoritma akan membuat kesimpulan bahwa ini adalah kucing.

### Terminologi Tambahan

Apa yang saya tunjukkan di sini adalah satu model pohon keputusan tertentu. Untuk memperkenalkan sedikit lebih banyak terminologi:

- **Simpul Akar**: Simpul paling atas di pohon.
- **Node Keputusan**: Semua oval di pohon, yang melihat fitur tertentu.
- **Simpul Daun**: Simpul di bagian bawah yang membuat prediksi.

### Perbedaan dengan Pohon Biologis

Jika Anda belum pernah melihat definisi ilmuwan komputer tentang pohon sebelumnya, mungkin tampak tidak intuitif bahwa akar pohon berada di atas dan daun pohon di bagian bawah. Mungkin salah satu cara untuk memikirkannya adalah ini lebih mirip dengan tanaman gantung dalam ruangan, itulah sebabnya akarnya berada di atas, dan daunnya cenderung jatuh ke dasar pohon.

## Contoh Pohon Keputusan Lainnya

Dalam slide ini, saya telah menunjukkan hanya satu contoh pohon keputusan. Berikut adalah beberapa lainnya. Ini adalah pohon keputusan yang berbeda untuk mencoba mengklasifikasikan kucing versus bukan kucing.

### Proses Keputusan Klasifikasi

- Di pohon ini, untuk membuat keputusan klasifikasi, Anda akan mulai lagi di simpul akar paling atas.
- Tergantung pada bentuk telinganya, Anda akan pergi ke kiri atau kanan.
- Jika bentuk telinga runcing, Anda melihat fitur kumis. 

#### Pohon Keputusan Berbeda

Hanya untuk bersenang-senang, inilah beberapa contoh pohon keputusan yang berbeda. Di antara pohon keputusan ini, beberapa akan lebih baik dan beberapa akan lebih buruk pada set pelatihan atau pada validasi silang dan set tes.

### Tujuan Algoritme Pembelajaran Pohon Keputusan

Tugas algoritme pembelajaran pohon keputusan adalah, dari semua pohon keputusan yang mungkin, untuk mencoba memilih salah satu yang mudah-mudahan berhasil dengan baik pada set pelatihan, dan juga idealnya menggeneralisasi dengan baik ke data baru seperti validasi silang dan set tes Anda juga.
