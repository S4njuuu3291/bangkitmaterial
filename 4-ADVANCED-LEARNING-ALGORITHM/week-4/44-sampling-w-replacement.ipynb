# Membangun Ensemble Pohon dengan Sampling dengan Penggantian

## Pendahuluan

Untuk membangun ensemble pohon, kita akan memerlukan teknik yang disebut **sampling with replacement**. Mari kita lihat apa artinya.

## Demonstrasi Sampling dengan Penggantian

Untuk mengilustrasikan cara kerja sampling dengan penggantian, saya akan menunjukkan demonstrasi sampling dengan penggantian menggunakan empat token yang berwarna merah, kuning, hijau, dan biru. Di sini saya memiliki empat token dengan warna merah, kuning, hijau, dan biru. Saya akan mendemonstrasikan seperti apa sampling dengan penggantian menggunakan token-token ini.

### Proses Sampling

1. **Persiapan**: Berikut adalah tas beludru hitam, yang kosong. Saya akan memasukkan keempat token ini ke dalam tas.
   
2. **Pengambilan Token**: Saya akan mengambil sampel sebanyak empat kali dengan penggantian dari tas ini. Ini berarti, saya akan mengocok tas, tidak bisa melihat saat saya mengambil, kemudian menarik satu token. Misalnya, token yang saya ambil adalah **hijau**.

3. **Penggantian**: Istilah "dengan penggantian" berarti bahwa jika saya mengambil token berikutnya, saya akan memasukkannya kembali ke dalam tas, mengocok lagi, lalu menarik satu lagi, misalnya, **kuning**. Kemudian saya mengembalikannya lagi. Itu adalah bagian dari penggantian.

4. **Proses Berlanjut**: Saya akan melanjutkan, misalnya, mengambil **biru** dan menggantinya lagi, lalu menarik satu lagi yang ternyata **biru** lagi. Urutan token yang saya dapatkan adalah **hijau, kuning, biru, biru**. 

### Catatan Penting

Perhatikan bahwa saya mendapatkan **biru** dua kali, dan tidak mendapatkan **merah** bahkan sekali. Jika Anda mengulangi prosedur sampling dengan penggantian ini beberapa kali, Anda mungkin mendapatkan urutan seperti **merah, kuning, merah, hijau**, atau **hijau, hijau, biru, merah**, atau bahkan **merah, biru, kuning, hijau**.

Penting untuk dicatat bahwa bagian "dengan penggantian" ini sangat kritis karena jika saya tidak mengganti token setiap kali saya mengambil sampel, maka jika saya mengeluarkan empat token dari tas yang berisi empat token, saya akan selalu mendapatkan keempat token yang sama. Oleh karena itu, mengganti token setelah saya mengeluarkannya setiap kali sangat penting untuk memastikan saya tidak hanya mendapatkan keempat token yang sama setiap kali.

## Aplikasi Sampling dengan Penggantian



Cara sampling dengan penggantian diterapkan untuk membangun ensemble pohon adalah sebagai berikut. Kita akan membangun beberapa set pelatihan acak yang semuanya sedikit berbeda dari set pelatihan asli kita.

1. **Menyiapkan Contoh**: Kita akan mengambil 10 contoh kucing dan anjing. Mari kita bayangkan kita menempatkan 10 contoh pelatihan ini dalam sebuah tas teoritis. Jangan memasukkan kucing atau anjing yang nyata ke dalam tas, itu terdengar tidak manusiawi. Namun, Anda dapat mengambil contoh pelatihan dan menempatkannya dalam tas teoritis jika Anda mau.

2. **Membuat Set Pelatihan Baru**: Menggunakan tas teoritis ini, kita akan membuat set pelatihan acak baru yang terdiri dari 10 contoh dengan ukuran yang sama persis dengan set data asli. Cara kita melakukannya adalah dengan menarik satu contoh pelatihan acak. Misalnya, kita mendapatkan contoh pelatihan ini.

3. **Pengembalian dan Pemilihan Ulang**: Setelah itu, kita mengembalikannya ke dalam tas, lalu menarik contoh pelatihan acak lainnya. Misalnya, kita mendapatkan contoh pelatihan lainnya. Anda terus melakukan ini hingga akhirnya mendapatkan 10 contoh pelatihan, beberapa di antaranya mungkin merupakan pengulangan.

4. **Catatan tentang Contoh Pelatihan**: Perhatikan bahwa set pelatihan ini tidak mengandung semua 10 contoh pelatihan asli, tetapi itu tidak masalah. Itu adalah bagian dari prosedur sampling dengan penggantian.



### Kesimpulan

Proses sampling dengan penggantian memungkinkan Anda untuk menyusun set pelatihan baru yang sedikit mirip dengan, tetapi juga cukup berbeda dari set pelatihan asli Anda.
