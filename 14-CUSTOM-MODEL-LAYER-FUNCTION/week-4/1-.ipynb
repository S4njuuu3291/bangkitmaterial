## Kustomisasi Arsitektur Model dengan Model Subclassing di Keras

Sebelumnya, Anda telah melihat bagaimana **kelas model Keras** dapat di-*subclass*, memungkinkan Anda untuk mengenkapsulasi model Anda dan memperluas bagaimana model dijalankan oleh kode Anda. Contohnya, Anda bisa membuat model yang kompleks dengan pendekatan **wide and deep** dalam satu baris kode di mana semua layer diimplementasikan di dalam kelas kustom, bukan di dalam kode utama Anda.



### Rekap Keuntungan Menggunakan Model Subclassing

Dengan menggunakan **model subclassing**, Anda tetap memanfaatkan **kelas model** yang sama seperti ketika membuat model menggunakan **Sequential** atau **Functional API**. Karena Anda memperluas kelas model asli, Anda dapat memanfaatkan berbagai fungsionalitas bawaan di dalamnya, termasuk fungsi **model.fit**, **evaluate**, dan **predict**, yang sudah ada dan tidak perlu di-*reimplementasi*. Selain itu, API untuk **penyimpanan** dan **serialisasi** tersedia, seperti menyimpan model atau sekumpulan **weights** tertentu.

Fitur-fitur seperti **model.summary** dan **plot_model** juga bisa digunakan pada model kustom Anda, sehingga Anda dapat merangkum atau memvisualisasikan arsitektur model dengan mudah. Bagi model yang lebih kompleks, seperti arsitektur **directed acyclic graphs** (DAG), pendekatan subclassing ini memungkinkan Anda membuat desain jaringan yang fleksibel.

### Keterbatasan pada API Sequential dan Functional



Dalam beberapa kasus arsitektur jaringan, seperti **MobileNet** dan **Inception**, yang dirancang untuk memproses data dari input hingga output secara langsung tanpa ada umpan balik atau **looping**, Sequential dan Functional API sudah cukup baik. Namun, untuk jaringan yang lebih dinamis, misalnya yang melibatkan **rekursi** atau perubahan struktur selama pelatihan atau inferensi, API ini menjadi lebih sulit digunakan. Dengan **Sequential**, hal ini hampir mustahil; dengan Functional API, mungkin memungkinkan tetapi membutuhkan banyak pengkodean tambahan. **Model subclassing** menawarkan cara yang lebih fleksibel untuk mencapai skenario kompleks ini.

### Keuntungan Menggunakan Model Subclassing



1. **Kemudahan Adaptasi**: Model subclassing memperluas pendekatan alami yang mungkin sudah Anda gunakan dalam membangun model menggunakan Sequential atau Functional API, sehingga Anda tidak perlu mempelajari keterampilan baru yang sangat berbeda.
   
2. **Kompatibilitas dengan Kode Lama**: Anda masih bisa menggunakan kode model yang sudah dibuat dengan Functional atau Sequential API, sehingga investasi waktu yang sudah dikeluarkan tetap bermanfaat dan bisa digunakan dalam proyek mendatang.

3. **Modularitas**: Dengan arsitektur subclass, Anda bisa merancang model secara modular. Misalnya, jika ingin bereksperimen dengan menambah atau mengurangi layer pada suatu bagian dari model, Anda bisa melakukannya tanpa harus mengubah keseluruhan kode. Anda hanya perlu menyesuaikan modul tertentu saja.

4. **Fleksibilitas dalam Aliran Data**: Anda bisa mengubah aliran data dalam jaringan. Selain alur dari layer atas ke layer bawah, Anda bisa menciptakan cabang-cabang atau loop dalam jaringan Anda.

Model subclassing memberikan kemampuan untuk membuat arsitektur model yang fleksibel dan kuat, memungkinkan eksperimen lebih cepat serta desain jaringan yang unik dan kompleks.
