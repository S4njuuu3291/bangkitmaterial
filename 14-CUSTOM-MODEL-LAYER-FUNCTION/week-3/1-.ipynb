## Kustomisasi Perilaku Layer dalam TensorFlow

Ada beberapa cara untuk menyesuaikan perilaku **layer** dalam **TensorFlow**. Cara pertama dan paling mudah, jika Anda hanya memerlukan fungsionalitas dasar, adalah dengan tidak membuat custom layer sama sekali, melainkan menggunakan **lambda layer**. Lambda layer adalah jenis layer yang memungkinkan Anda menjalankan kode secara arbitrer. Seperti yang dijelaskan, tujuan dari lambda layer adalah untuk mengeksekusi fungsi tertentu dalam model berbasis **Sequential** atau **Functional API**.

Lambda layer ini sangat cocok untuk eksperimen cepat atau sesuatu yang sederhana. Contoh lambda layer paling sederhana mungkin terlihat seperti ini: Anda akan menentukan nilai lambda, misalnya `x`, dan kemudian nilai tersebut akan dipetakan, dalam hal ini ke nilai absolut `x`. Sebagai contoh, nilai `-1` akan dipetakan menjadi `1` oleh lambda layer ini.

### Contoh Penggunaan dalam Model

Mari kita pertimbangkan arsitektur model pada dataset **Fashion MNIST** yang telah sering kita gunakan. Dalam model ini, terdapat layer **dense** yang diaktifkan oleh **ReLU**. Fungsi ReLU secara efektif menghilangkan output negatif dari layer tersebut, sehingga hanya nilai positif yang mengalir ke layer berikutnya. Hal ini menghindari nilai negatif dan positif yang akan saling meniadakan.

#### Hasil Pelatihan dengan ReLU

Setelah melatih jaringan selama lima *epoch*, hasilnya cukup baik, dengan akurasi sebesar 98,6% pada set pelatihan dan 97,5% pada set validasi. Waktu pelatihan juga sekitar empat detik per langkah.

#### Hasil Tanpa ReLU

Jika kita menghilangkan ReLU, akan terlihat dampaknya pada pelatihan dan performa model. Akurasi menurun menjadi 92,2% pada set pelatihan dan 91,5% pada set validasi, dengan waktu pelatihan yang tetap sama.

#### Menggunakan Fungsi Absolut dengan Lambda Layer

Dengan mengubah arsitektur model dan mengganti ReLU dengan fungsi absolut pada lambda layer, kita bisa melihat sedikit peningkatan akurasi, yakni 98,7% pada set pelatihan dan 97,5% pada set validasi. Waktu pelatihan pun sedikit lebih cepat, dengan sekitar tiga detik per langkah.
