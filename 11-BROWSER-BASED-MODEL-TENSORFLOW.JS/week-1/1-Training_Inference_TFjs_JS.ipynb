# **Selamat datang di kursus TensorFlow.js**


### **Pengenalan TensorFlow.js**

- The goal is twofold:
  1. **Mudah untuk digunakan** dengan **high-level API** yang ramah pengguna.
  2. **Akses API tingkat rendah** untuk pemrograman langsung.

- TensorFlow.js dirancang untuk bekerja **di browser** dan **Node.js server**.

---

### **Koneksi dengan Keras API**

Sampai saat ini, kamu telah memprogram **convolutional neural networks** dan **deep neural networks** menggunakan **Keras API** di **TensorFlow**. **Layers API** di TensorFlow.js terlihat dan terasa sangat mirip dengan **Keras**. Jadi, apa yang telah kamu pelajari sebelumnya dapat langsung diterapkan di sini, meskipun dengan sedikit perbedaan sintaks karena menggunakan **JavaScript** daripada **Python**.

---



### **Core APIs**

- **Core APIs** adalah API tingkat rendah yang dirancang untuk bekerja dengan format **TensorFlow saved model**.
- Di TensorFlow 2.0, format model ini standar dan dapat digunakan di **Python APIs**, **JavaScript APIs**, dan bahkan **TensorFlow Lite** untuk perangkat **mobile** dan **embedded**.
  
  - Core API bekerja di **browser** dan dapat memanfaatkan **WebGL** untuk mempercepat proses training dan inference.

---

### **Penggunaan di Node.js**

- Pada **Node.js**, kamu bisa membangun aplikasi **server-side** atau **terminal** menggunakan TensorFlow.js.
- Aplikasi tersebut dapat memanfaatkan **CPUs**, **GPUs**, dan **TPUs** tergantung pada perangkat yang tersedia.

---

### **Membangun Neural Network Sederhana**

- Kita akan membangun **neural network** yang sangat sederhana, yang berfungsi untuk mencocokkan dua set angka dengan rumus hubungan:  
  `y = 2 * x - 1`.

- Proses ini akan dilakukan menggunakan **JavaScript** di **browser**.

---

### **Ringkasan:**
- **TensorFlow.js** memungkinkan pengembangan model machine learning **di browser** dan **Node.js server**.
- Kamu bisa menggunakan API **tingkat tinggi** untuk kemudahan, atau **API tingkat rendah** untuk kontrol lebih dalam.
- Model yang dibangun dapat dijalankan menggunakan **CPUs**, **GPUs**, dan **TPUs**.
