```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
y_train.shape[0]
```

```python
plt.scatter(x_train, y_train,marker="+",color="red")
plt.title("Harga Rumah")
plt.xlabel("Ukuran")
plt.ylabel("Harga (dalam $1000)")
plt.show()
```

```python
w = 200
b = 100

def hitung_output(x,w,b):
    fw_b = np.zeros(x.shape[0])
    for i in range(x.shape[0]):
        fw_b[i] = w * x[i] + b
    return fw_b
```

```python
fw_b = hitung_output(x_train,w,b)
fw_b
```

```python
plt.plot(x_train,fw_b,color="red",marker="o")
plt.scatter(x_train,y_train,marker="o",color="blue")
plt.show()
```

```python
size = 1.2

harga = w*size + b
print(f"Harga rumah dengan ukuran {size} adalah ${harga:.0f} thousand dollars ")
```
