```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Data awal

# Menambahkan 50 data baru dengan variasi acak
x_additional = np.linspace(3, 52, 150)  # Membuat 50 data baru untuk x (dari 3 hingga 52)

# Menghasilkan data y dengan variasi acak
# Misalnya, menggunakan sedikit noise acak berdasarkan x
y_additional = 200 * x_additional + 100 + np.random.normal(0, 150, 150)  # Noise acak ditambahkan

# Menggabungkan data awal dengan data tambahan
x_train = (x_additional)
y_train = (y_additional)

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
y_train
```

```python
plt.scatter(x_train, y_train,marker="+",color="red")
plt.title("Harga Rumah")
plt.xlabel("Ukuran")
plt.ylabel("Harga (dalam $1000)")
plt.show()
```

```python
w = 200
b = 100

def hitung_output(x,w,b):
    fw_b = np.zeros(x.shape[0])
    for i in range(x.shape[0]):

        fw_b[i] = w * x[i] + b
    return fw_b
```

```python
fw_b = hitung_output(x_train,w,b)
fw_b
```

```python
plt.plot(x_train,fw_b,color="red",marker="o")
plt.scatter(x_train,y_train,marker="o",color="blue")
plt.show()
```

```python
size = 200

harga = w*size + b
print(f"Harga rumah dengan ukuran {size} adalah ${harga:.0f} thousand dollars ")
```

## Cost Function

```python
# Fungsi untuk menghitung output prediksi (misal regresi linear)
def hitung_output(x, w, b):
    return np.dot(x, w) + b

# Fungsi MSE yang diperbaiki
def mse(x, y, w, b):
    y_hat = hitung_output(x, w, b)  # Menghitung prediksi
    error = np.mean((y_hat - y) ** 2)  # Menggunakan operasi vektor dan menghitung rata-rata
    return error

j = mse(x_train, y_train, w, b)
j
```
