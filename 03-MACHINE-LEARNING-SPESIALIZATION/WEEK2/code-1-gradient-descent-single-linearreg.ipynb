```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

n_samples = 600

X = np.random.rand(n_samples,1)*50

w_true = 21
b_true = 4

noise = np.random.rand(n_samples,1)*100
y = w_true*X + b_true + noise
```

```python
def mse(X,y,w,b):
    error = 0
    for i in range(len(X)):
        y_pred = w*X[i]+b
        error = error + (y_pred - y[i])**2
    error/=(2*len(X))
    return error
```

```python
mse(X,y,w_true,b_true)
```

```python
def hitung_gradien(X,y,w,b):
    m = len(X)
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        dj_dw += ((w*X[i]+b)-y[i])*X[i]
        dj_db += ((w*X[i]+b)-y[i])
    dj_dw/=m
    dj_db/=m
    return dj_dw,dj_db
```

```python
hitung_gradien(X,y,w_true,b_true)
```

```python
def gradient_descent(X,y,w,b,iterasi,alpha):
    m = len(X)
    w_baru = w
    b_baru = b

    loss_hist = []
    loss = 0
    for i in range(iterasi):
        dj_dw,dj_db = hitung_gradien(X,y,w_baru,b_baru)
        w_baru = w_baru - alpha*dj_dw
        b_baru = b_baru - alpha*dj_db
        loss = mse(X,y,w_baru,b_baru)
        loss_hist.append(loss)
        if i%1 == 0:
            print(f"Iterasi {i+1}: Loss = {loss}")
    print(f"Final Loss : {loss}, w = {w_baru}, b = {b_baru}")
    return loss_hist,w_baru,b_baru
```

```python
w_init = 15
b_init = 50
iter = 1000
loss_hist,w,b=gradient_descent(X,y,w_init,b_init,iter,0.0001)
```
