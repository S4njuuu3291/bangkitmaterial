# Kurva Pembelajaran

Kurva pembelajaran adalah cara untuk memahami bagaimana kinerja algoritme pembelajaran sebagai fungsi dari jumlah pengalaman yang dimilikinya. Dalam hal ini, pengalaman mengacu pada **jumlah contoh pelatihan** yang dimiliki algoritme.

## Memplot Kurva Pembelajaran

Mari kita memplot kurva pembelajaran untuk model yang sesuai dengan fungsi kuadrat polinomial orde kedua. Kita akan memplot dua hal berikut:
- **J_CV**: Kesalahan validasi silang.
- **J_train**: Kesalahan pelatihan.

### Sumbu yang Digunakan
- **Sumbu horizontal**: `m_train`, yaitu ukuran set pelatihan atau jumlah contoh pelatihan.
- **Sumbu vertikal**: Kesalahan (`J_CV` atau `J_train`).

## Mengapa Kesalahan Pelatihan Meningkat?
Misalkan Anda memiliki hanya satu contoh pelatihan, maka model kuadrat mungkin akan dengan mudah mendapatkan **kesalahan pelatihan nol**. Namun, seiring bertambahnya jumlah contoh pelatihan (dua, tiga, dan seterusnya), semakin sulit bagi model untuk menyesuaikan semuanya secara sempurna. 

## Perbandingan J_CV dan J_train
Biasanya, **J_CV** akan lebih tinggi daripada **J_train** karena parameter dioptimalkan dengan menggunakan set pelatihan. **J_train** bisa lebih rendah pada set pelatihan, tetapi pada set validasi, model cenderung memiliki performa yang lebih rendah.

## Kurva Pembelajaran untuk Kasus Bias Tinggi (Underfitting)

Kurva belajar untuk bias tinggi (underfitting) biasanya akan menunjukkan pola sebagai berikut:
- **Kesalahan pelatihan** akan naik dan kemudian **mendatar** setelah beberapa saat.
- **Kesalahan validasi silang** akan turun dan mendatar, dengan nilai yang lebih tinggi dari **J_train**.



Pada model bias tinggi seperti ini, mendapatkan lebih banyak data pelatihan **tidak akan membantu** meningkatkan performa model. Ini karena model terlalu sederhana untuk menangkap pola data yang kompleks.

## Kurva Pembelajaran untuk Kasus Varians Tinggi (Overfitting)

Sekarang, mari kita lihat kurva pembelajaran untuk model dengan varians tinggi (overfitting). Pada varians tinggi:
- **Kesalahan pelatihan** akan lebih rendah, tetapi **kesalahan validasi silang** jauh lebih tinggi daripada **J_train**.



Perbedaan yang besar antara **J_train** dan **J_CV** mengindikasikan bahwa model terlalu pas dengan data pelatihan namun gagal melakukan generalisasi pada data baru.

### Mengatasi Varians Tinggi dengan Menambah Data
Pada varians tinggi, **menambah lebih banyak data pelatihan** bisa sangat membantu. Hal ini akan menurunkan kesalahan validasi silang dan mendekatkan performa model pada kinerja optimal (misalnya, **kinerja tingkat manusia**).

---

## Menganalisis Kurva Pembelajaran

Jika Anda membangun aplikasi pembelajaran mesin, Anda dapat **merencanakan kurva pembelajaran** untuk memahami apakah model Anda mengalami bias tinggi atau varians tinggi. Caranya adalah dengan mengambil subset yang berbeda dari set pelatihan Anda dan memplot **J_train** dan **J_CV** untuk melihat pola yang terbentuk.

Namun, satu kelemahan dari metode ini adalah bahwa memplot banyak kurva belajar cukup mahal secara komputasi, terutama jika Anda harus melatih beberapa model dengan subset pelatihan yang berbeda.

## Kesimpulan: Bias vs Varians
Untuk rekap:
- Jika model Anda **memiliki bias tinggi**, menambah data pelatihan **tidak akan membantu**.
- Jika model Anda **memiliki varians tinggi**, menambah data pelatihan **dapat sangat membantu** dalam menurunkan kesalahan validasi silang dan meningkatkan performa.

Jika algoritme Anda mengalami varians tinggi, terus menambah lebih banyak data pelatihan dapat mendekatkan kinerja model pada **kinerja optimal**. Sebaliknya, jika Anda melihat bias tinggi, Anda mungkin perlu **menggunakan model yang lebih kompleks** sebelum menambah data pelatihan.
