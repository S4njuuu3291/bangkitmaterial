# Kelemahan Menggunakan Satu Pohon Keputusan

## Pendahuluan

Salah satu kelemahan menggunakan satu pohon keputusan adalah bahwa pohon keputusan tersebut dapat sangat sensitif terhadap perubahan kecil dalam data. Salah satu solusi untuk membuat algoritma ini kurang sensitif atau lebih robust adalah dengan membangun bukan hanya satu pohon keputusan, tetapi banyak pohon keputusan. Kita menyebut ini sebagai **ensemble pohon**. Mari kita lihat lebih dekat.



## Contoh

Dengan contoh yang telah kita gunakan, fitur terbaik untuk membagi pada simpul akar ternyata adalah bentuk telinga, yang menghasilkan dua subset data dan kemudian membangun sub-pohon lebih lanjut pada dua subset tersebut. Namun, jika Anda hanya mengambil satu dari sepuluh contoh dan mengubahnya menjadi kucing yang berbeda, sehingga bukannya memiliki telinga runcing, wajah bulat, dan kumis tidak ada, kucing baru ini memiliki telinga floppy, wajah bulat, dan kumis hadir. Dengan hanya mengubah satu contoh pelatihan, fitur dengan informasi tertinggi untuk dibagi menjadi fitur kumis, bukan fitur bentuk telinga.

### Akibat Perubahan

Akibatnya, subset data yang Anda dapatkan di sub-pohon kiri dan kanan menjadi sangat berbeda. Ketika Anda terus menjalankan algoritma pembelajaran pohon keputusan secara rekursif, Anda membangun sub-pohon yang sangat berbeda di kiri dan kanan. Fakta bahwa mengubah hanya satu contoh pelatihan menyebabkan algoritma menghasilkan pembagian yang berbeda di akar dan dengan demikian menghasilkan pohon yang sama sekali berbeda, menunjukkan bahwa algoritma ini tidak cukup robust.

## Mengapa Ensemble Pohon?

Itulah sebabnya, ketika Anda menggunakan pohon keputusan, Anda sering mendapatkan hasil yang jauh lebih baik, yaitu prediksi yang lebih akurat jika Anda melatih bukan hanya satu pohon keputusan, tetapi sejumlah besar pohon keputusan yang berbeda. Inilah yang kita sebut sebagai **ensemble pohon**, yang berarti sekumpulan banyak pohon.

### Proses Pengambilan Keputusan

Mari kita lihat bagaimana kita dapat membangun ensemble pohon ini. Jika Anda memiliki ensemble tiga pohon, masing-masing dari pohon ini mungkin merupakan cara yang layak untuk mengklasifikasikan kucing versus bukan kucing. Jika Anda memiliki contoh uji baru yang ingin Anda klasifikasikan, maka yang perlu Anda lakukan adalah menjalankan ketiga pohon ini pada contoh baru Anda dan meminta mereka untuk memberikan suara pada prediksi akhir.

## Contoh Prediksi

Contoh uji ini memiliki telinga runcing, bentuk wajah yang tidak bulat, dan kumis hadir, jadi:
- **Pohon Pertama:** Menginformasikan bahwa ini adalah kucing.
- **Pohon Kedua:** Menginformasikan bahwa ini bukan kucing.
- **Pohon Ketiga:** Menginformasikan bahwa ini adalah kucing.



Ketiga pohon ini telah memberikan prediksi yang berbeda, jadi kita akan meminta mereka untuk memberikan suara. Suara mayoritas dari prediksi di antara ketiga pohon ini adalah kucing. Dengan demikian, prediksi akhir dari ensemble pohon ini adalah bahwa ini adalah kucing, yang kebetulan merupakan prediksi yang benar.

## Keunggulan Ensemble

Alasan kita menggunakan ensemble pohon adalah dengan memiliki banyak pohon keputusan dan meminta mereka untuk memberikan suara, hal ini membuat algoritma keseluruhan Anda kurang sensitif terhadap apa yang mungkin dilakukan oleh setiap pohon tunggal. Setiap pohon hanya memberikan satu suara dari tiga atau satu suara dari banyak suara yang berbeda, sehingga membuat algoritma Anda lebih robust.
