# Perbandingan antara Decision Trees, Tree Ensembles, dan Neural Networks

## Pendahuluan
Baik decision trees (pohon keputusan) maupun tree ensembles (ensambel pohon keputusan) serta neural networks (jaringan saraf) adalah algoritma pembelajaran yang sangat kuat dan efektif. Kapan sebaiknya Anda memilih salah satu daripada yang lain? Mari kita lihat beberapa kelebihan dan kekurangan masing-masing.

## Decision Trees dan Tree Ensembles
Decision trees dan tree ensembles biasanya bekerja dengan baik pada data tabular, juga disebut sebagai data terstruktur. Ini berarti jika dataset Anda terlihat seperti spreadsheet besar, maka decision trees layak dipertimbangkan. 

### Contoh Kasus
Misalnya, dalam aplikasi prediksi harga rumah, kita memiliki dataset dengan fitur yang sesuai dengan ukuran rumah, jumlah kamar tidur, jumlah lantai, dan usia rumah. Tipe data ini disimpan dalam spreadsheet dengan fitur bernilai kategorikal atau kontinu, dan cocok untuk tugas klasifikasi atau regresi di mana Anda mencoba memprediksi kategori diskrit atau angka.

Semua masalah ini adalah hal yang dapat ditangani dengan baik oleh decision trees. Sebaliknya, saya tidak akan merekomendasikan penggunaan decision trees dan tree ensembles pada data tidak terstruktur, seperti gambar, video, audio, dan teks, yang kurang mungkin disimpan dalam format spreadsheet.

### Keuntungan Decision Trees
Salah satu keuntungan besar dari decision trees dan tree ensembles adalah kecepatan pelatihan yang tinggi. Anda mungkin ingat diagram dari minggu lalu di mana kita membahas tentang siklus iteratif pengembangan machine learning. Jika model Anda memerlukan banyak waktu untuk dilatih, maka itu membatasi seberapa cepat Anda dapat melalui siklus ini dan meningkatkan kinerja algoritma Anda.

Karena decision trees, termasuk tree ensembles, cenderung cepat untuk dilatih, ini memungkinkan Anda untuk masuk ke siklus ini lebih cepat dan mungkin secara efisien meningkatkan kinerja algoritma pembelajaran Anda.

### Interpretabilitas
Akhirnya, decision trees kecil mungkin dapat diinterpretasikan oleh manusia. Jika Anda melatih hanya satu decision tree dan pohon keputusan itu hanya memiliki beberapa lusin node, Anda mungkin bisa mencetak pohon keputusan untuk memahami bagaimana ia membuat keputusan. Namun, interpretabilitas decision trees terkadang agak dilebih-lebihkan, karena ketika Anda membangun ensambel 100 pohon dan masing-masing pohon tersebut memiliki ratusan node, maka melihat ensambel tersebut untuk menentukan apa yang dilakukannya menjadi sulit dan mungkin memerlukan teknik visualisasi terpisah.

## Pilihan untuk Menggunakan XGBoost
Jika Anda memutuskan untuk menggunakan decision tree atau tree ensemble, saya mungkin akan menggunakan XGBoost untuk sebagian besar aplikasi yang akan saya kerjakan. Satu kelemahan kecil dari tree ensemble adalah bahwa ia sedikit lebih mahal dibandingkan dengan satu pohon keputusan. Jika Anda memiliki anggaran komputasi yang sangat terbatas, Anda mungkin menggunakan satu pohon keputusan, tetapi di luar pengaturan tersebut, saya hampir selalu akan menggunakan tree ensemble dan menggunakan XGBoost secara khusus.

## Neural Networks
Berbeda dengan decision trees dan tree ensembles, neural networks bekerja dengan baik pada semua jenis data, termasuk data tabular atau terstruktur, serta data tidak terstruktur. Selain itu, neural networks juga dapat menangani data campuran yang mencakup komponen terstruktur dan tidak terstruktur. 

### Kelebihan Neural Networks
Meskipun pada data terstruktur tabular, neural networks dan decision trees sering kali bersaing secara kompetitif, pada data tidak terstruktur, seperti gambar, video, audio, dan teks, neural networks benar-benar menjadi algoritma pilihan dan bukan decision tree atau tree ensemble.

Namun, di sisi lain, neural networks mungkin lebih lambat daripada decision tree. Jaringan saraf yang besar dapat memerlukan waktu lama untuk dilatih. Manfaat lain dari neural networks termasuk kemampuannya untuk bekerja dengan transfer learning, yang sangat penting karena untuk banyak aplikasi kita hanya memiliki dataset kecil. Kemampuan untuk menggunakan transfer learning dan melakukan pra-pelatihan pada dataset yang jauh lebih besar adalah kunci untuk mendapatkan kinerja yang kompetitif.



### Kolaborasi Model
Terakhir, jika Anda membangun sistem yang terdiri dari beberapa model machine learning yang bekerja bersama, mungkin lebih mudah untuk menggabungkan dan melatih beberapa neural networks daripada beberapa decision trees. Alasan untuk ini cukup teknis, dan Anda tidak perlu khawatir tentang itu untuk tujuan kursus ini. Namun, ini berkaitan dengan kenyataan bahwa bahkan ketika Anda menggabungkan beberapa neural networks, Anda dapat melatih semuanya bersama menggunakan gradient descent, sedangkan untuk decision trees Anda hanya dapat melatih satu decision tree pada satu waktu.
