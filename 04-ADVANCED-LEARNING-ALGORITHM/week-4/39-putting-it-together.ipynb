# Proses Membangun Pohon Keputusan

## Kriteria Informasi Gain

Kriteria **information gain** memungkinkan kita untuk memilih satu fitur untuk membagi satu node. Mari kita gunakan ini di beberapa tempat melalui pohon keputusan untuk mengetahui cara membangun pohon keputusan besar dengan beberapa node.

## Langkah-Langkah Membangun Pohon Keputusan

Proses membangun pohon keputusan dimulai dengan semua contoh pelatihan di node akar pohon. Berikut adalah langkah-langkahnya:

1. **Menghitung Information Gain**  
   Hitung information gain untuk semua fitur yang mungkin dan pilih fitur yang memberikan information gain tertinggi.

2. **Membagi Dataset**  
   Setelah memilih fitur ini, bagi dataset menjadi dua subset sesuai dengan fitur yang dipilih. Buat cabang kiri dan kanan dari pohon, dan kirim contoh pelatihan ke cabang kiri atau kanan, tergantung pada nilai fitur tersebut untuk contoh itu.

3. **Pengulangan Proses Pembagian**  
   Ulangi proses pembagian ini pada cabang kiri dan kanan dari pohon sampai kriteria penghentian terpenuhi.



## Kriteria Penghentian

Kriteria penghentian dapat berupa:

- Ketika node 100% terdiri dari satu kelas.
- Ketika entropi mencapai nol.
- Ketika pembagian lebih lanjut menyebabkan pohon melebihi kedalaman maksimum yang telah ditetapkan.
- Ketika information gain dari pembagian tambahan kurang dari ambang batas tertentu.
- Ketika jumlah contoh dalam node di bawah ambang batas tertentu.

Ulangi proses pembagian sampai kriteria penghentian yang telah Anda pilih terpenuhi.

## Ilustrasi Proses

Misalkan kita memulai dengan semua contoh di node akar dan menghitung information gain untuk semua fitur. Kita mungkin memutuskan bahwa **bentuk telinga** adalah fitur terbaik untuk dibagi. Berdasarkan keputusan itu, kita membuat cabang kiri dan kanan dan mengirim subset data dengan bentuk telinga runcing dan telinga floppy ke cabang kiri dan kanan.

### Cabang Kiri

Mari kita fokus pada cabang kiri di mana terdapat lima contoh. Kita periksa kriteria penghentian:

- Jika node tersebut terdiri dari satu kelas (semua kucing atau semua anjing), kriteria penghentian terpenuhi, dan kita buat leaf node yang memprediksi kelas tersebut.

### Cabang Kanan

Untuk cabang kanan, kita memeriksa lima contoh lagi. Jika kriteria penghentian tidak terpenuhi, kita akan terus membagi:

1. Hitung information gain untuk semua fitur.
2. Pilih fitur dengan information gain tertinggi untuk melakukan pembagian.
3. Ulangi hingga kriteria penghentian terpenuhi.



## Rekursi dalam Algoritma

Setelah kita memutuskan fitur yang akan dibagi di node akar, cara kita membangun subtree kiri adalah dengan membangun pohon keputusan pada subset contoh yang lebih kecil. Ini adalah contoh dari **algoritma rekursif**, di mana kita membangun pohon keputusan secara keseluruhan dengan membangun subtree yang lebih kecil.

## Memilih Parameter Kedalaman Maksimum

Ketika memilih parameter kedalaman maksimum, ada beberapa pilihan yang mungkin. Umumnya:

- Semakin besar kedalaman maksimum, semakin besar pohon keputusan yang ingin Anda bangun. Ini mirip dengan menyesuaikan polinomial derajat tinggi, yang dapat menyebabkan overfitting jika model terlalu kompleks.

## Kriteria Tambahan untuk Penghentian

Kriteria lain untuk menghentikan pembagian adalah:

- Jika information gain dari pembagian tambahan hanya menghasilkan pengurangan kecil dalam entropi.
- Jika jumlah contoh dalam node di bawah ambang batas tertentu.

## Melakukan Prediksi

Setelah membangun pohon keputusan, jika Anda ingin membuat prediksi, Anda dapat mengikuti prosedur dengan mengambil contoh baru, mulai dari akar pohon, dan terus mengikuti keputusan hingga mencapai leaf node, yang akan memberikan prediksi akhir.
