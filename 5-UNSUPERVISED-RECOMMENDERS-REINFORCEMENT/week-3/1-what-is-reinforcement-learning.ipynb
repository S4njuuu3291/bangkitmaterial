## Menggunakan Reinforcement Learning untuk Mengendalikan Helicopter

Bagaimana cara membuat helicopter terbang sendiri menggunakan reinforcement learning? Tugasnya adalah, diberikan posisi helicopter, untuk memutuskan bagaimana menggerakkan joystick kontrol. Dalam reinforcement learning, kami menyebut posisi, orientasi, dan kecepatan helicopter sebagai **state (s)**. Tugasnya adalah menemukan fungsi yang memetakan dari state helicopter ke **action (a)**, yang berarti seberapa jauh harus mendorong kedua joystick untuk menjaga helicopter tetap seimbang di udara dan terbang tanpa jatuh.

Satu cara untuk mencoba menyelesaikan masalah ini adalah dengan menggunakan **supervised learning**. Ternyata, ini bukan pendekatan yang bagus untuk penerbangan helicopter otonom. Namun, Anda bisa berkata, "Jika kita bisa mendapatkan banyak pengamatan dari state dan mungkin seorang pilot manusia ahli memberi tahu kita apa action terbaik (y) yang harus diambil." Anda kemudian bisa melatih **neural network** menggunakan supervised learning untuk langsung mempelajari pemetaan dari state (s) yang saya sebut x, ke action (a) yang saya sebut label y.

Tetapi ternyata, saat helicopter bergerak melalui udara, sangat ambigu untuk menentukan tindakan yang tepat. Apakah Anda harus sedikit miring ke kiri atau lebih miring ke kiri lagi, atau meningkatkan ketegangan helicopter sedikit atau banyak? Sangat sulit untuk mendapatkan dataset dari x dan tindakan ideal y. Itulah sebabnya untuk banyak tugas mengendalikan robot seperti helicopter dan robot lainnya, pendekatan supervised learning tidak bekerja dengan baik, dan kami menggunakan reinforcement learning sebagai gantinya.

## Fungsi Reward

Input kunci untuk reinforcement learning adalah sesuatu yang disebut **reward** atau **reward function** yang memberi tahu helicopter kapan ia melakukan dengan baik dan kapan ia melakukan dengan buruk. Cara saya memahami fungsi reward adalah mirip dengan melatih anjing. Ketika saya tumbuh dewasa, keluarga saya memiliki seekor anjing, dan itu adalah tugas saya untuk melatih anjing tersebut agar bersikap baik.

### Melatih Anjing dengan Reinforcement Learning

Jadi, bagaimana Anda mendapatkan puppy untuk berperilaku baik? Anda tidak bisa banyak mendemonstrasikan kepada puppy. Sebaliknya, Anda membiarkannya melakukan hal-hal yang ingin dilakukannya, dan setiap kali ia melakukan sesuatu yang baik, Anda berkata, "Good dog!" Dan setiap kali ia melakukan sesuatu yang buruk, Anda berkata, "Bad dog!" Dan kemudian, semoga, ia belajar dengan sendirinya untuk melakukan lebih banyak hal baik dan lebih sedikit hal buruk.

Melatih dengan algoritma reinforcement learning seperti itu. Ketika helicopter terbang dengan baik, Anda berkata, "Good helicopter!" dan jika ia melakukan sesuatu yang buruk seperti jatuh, Anda berkata, "Bad helicopter!" Dan kemudian tugas algoritma reinforcement learning adalah mencari cara agar lebih banyak hasil baik dan lebih sedikit hasil buruk untuk helicopter.



## Fleksibilitas Reinforcement Learning

Salah satu alasan mengapa reinforcement learning sangat kuat adalah karena Anda harus memberi tahu apa yang harus dilakukan, bukan bagaimana cara melakukannya. Menentukan fungsi reward daripada tindakan optimal memberi Anda lebih banyak fleksibilitas dalam merancang sistem. Secara konkret, untuk menerbangkan helicopter, setiap kali ia terbang dengan baik, Anda dapat memberinya reward +1 setiap detik ia terbang dengan baik. Dan mungkin setiap kali ia terbang buruk, Anda dapat memberinya reward negatif, atau jika ia jatuh, Anda dapat memberinya reward negatif yang sangat besar, seperti -1,000. Ini akan mendorong helicopter untuk menghabiskan lebih banyak waktu terbang dengan baik dan semoga tidak pernah jatuh.

## Aplikasi Reinforcement Learning

Hari ini, reinforcement learning telah berhasil diterapkan dalam berbagai aplikasi mulai dari mengendalikan robot. Bahkan, di laboratorium praktik akhir pekan ini, Anda akan menerapkan algoritma reinforcement learning untuk mendaratkan **lunar lander** dalam simulasi. Jadi, apa cara terbaik untuk menjadwalkan perdagangan Anda dari waktu ke waktu sehingga Anda dapat menjual saham yang ingin Anda jual dan semoga mendapatkan harga terbaik untuknya?

Akhirnya, ada juga banyak aplikasi reinforcement learning dalam permainan, mulai dari **catur**, **checkers**, **bridge**, hingga **go**, serta banyak **video games**.

## Kesimpulan

Jadi, itulah reinforcement learning. Meskipun reinforcement learning tidak digunakan sebanyak supervised learning, masih ada beberapa aplikasi yang menggunakan konsep ini saat ini. Ide kunci adalah, daripada Anda perlu memberi tahu algoritma apa output yang benar (y) untuk setiap input, semua yang perlu Anda lakukan adalah menentukan fungsi reward yang memberi tahu kapan ia melakukan dengan baik dan kapan ia melakukan dengan buruk. Tugas algoritma adalah secara otomatis mencari cara untuk memilih tindakan yang baik.
