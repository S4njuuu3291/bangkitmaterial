# Lunar Lander: Simulasi Pendaratan di Bulan

Lunar lander memungkinkan Anda untuk mendaratkan sebuah kendaraan simulasi di bulan. Ini seperti permainan video kecil yang menyenangkan dan telah digunakan oleh banyak peneliti di bidang **reinforcement learning**. Mari kita lihat lebih dalam mengenai aplikasi ini.

## Deskripsi Aplikasi

Dalam aplikasi ini, Anda mengendalikan sebuah lunar lander yang dengan cepat mendekati permukaan bulan. Tugas Anda adalah menembakkan **thrusters** pada waktu yang tepat untuk mendaratkannya dengan aman di landasan pendaratan. Berikut adalah dua skenario yang mungkin terjadi:

- **Pendaratan Sukses**: Lunar lander berhasil mendarat dengan menembakkan thrusters ke bawah serta ke kiri dan kanan untuk memposisikan dirinya di antara dua bendera kuning.
  
![Pendaratan Sukses](URL_GAMBAR_PENDARATAN_SUKSES)

- **Kecelakaan**: Jika algoritma pendaratan **reinforcement** tidak berfungsi dengan baik, lander dapat mengalami kecelakaan dan hancur saat menyentuh permukaan bulan.
  
![Kecelakaan](URL_GAMBAR_KECELAKAAN)

## Aksi yang Tersedia

Dalam aplikasi ini, Anda memiliki empat aksi yang mungkin pada setiap langkah waktu:

1. **Nothing**: Tidak melakukan apa-apa, sehingga gaya inersia dan gravitasi akan menarik Anda ke permukaan bulan.
2. **Left**: Menembakkan thruster kiri, yang akan mendorong lunar lander ke kanan.
3. **Main**: Menembakkan mesin utama yang mendorong ke bawah.
4. **Right**: Menembakkan thruster kanan, yang akan mendorong lunar lander ke kiri.

Untuk menyederhanakan, saya akan menyebut aksi-aksi ini sebagai `nothing`, `left`, `main`, dan `right`.

## Ruang Status

**Status** dari lunar lander terdiri dari beberapa variabel:

- **Posisi (X, Y)**: Jarak horizontal (kiri atau kanan) dan tinggi dari permukaan bulan.
- **Kecepatan (vx, vy)**: Kecepatan bergerak dalam arah horizontal dan vertikal.
- **Sudut (angle)**: Seberapa miring lunar lander ke kiri atau kanan.
- **Kecepatan Sudut (angular velocity, theta dot)**: Perubahan sudut dalam waktu.
- **Dua Variabel (l dan r)**: Menunjukkan apakah kaki kiri (`l`) atau kaki kanan (`r`) menyentuh tanah, dengan nilai biner (0 atau 1).

## Fungsi Reward

Fungsi reward untuk lunar lander dirancang sebagai berikut:

- Jika berhasil mendarat di landasan, reward antara 100 dan 140 diberikan, tergantung seberapa baik penerbangannya.
- Menerima reward tambahan untuk bergerak lebih dekat ke landasan dan penalti untuk menjauh.
- Jika lunar lander mengalami kecelakaan, penalti besar -100 diberikan.
- Untuk setiap kaki yang berhasil mendarat, reward +100 diberikan.
- Setiap kali mesin utama dinyalakan, penalti -0.3 dan -0.03 untuk thrusters kiri atau kanan.

Desain fungsi reward ini cukup kompleks dan mencerminkan pemikiran mendalam tentang perilaku yang diinginkan.

## Tujuan Pembelajaran

Tujuan dari aplikasi lunar lander adalah untuk mempelajari **policy** (π) yang, ketika diberikan sebuah status S, memilih aksi A = π(S) untuk memaksimalkan total reward yang didiskontokan. Biasanya, kita menggunakan nilai besar untuk gamma (γ), yaitu 0.985, yang cukup mendekati satu.

Jika Anda berhasil mempelajari policy π yang efektif, Anda akan berhasil mendaratkan lunar lander. Sekarang kita siap untuk mengembangkan algoritma pembelajaran yang akan menggunakan **deep learning** atau **neural networks** untuk menentukan policy pendaratan lunar lander.
