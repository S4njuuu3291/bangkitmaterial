# Penyelesaian Formalisme Pembelajaran Penguatan

Untuk menyelesaikan formalism pembelajaran penguatan, alih-alih melihat sesuatu yang rumit seperti helikopter atau robot anjing, kita dapat menggunakan contoh yang disederhanakan yang terinspirasi dari Mars rover. Contoh ini diadaptasi dari karya Profesor Stanford, Emma Branskill, dan salah satu kolaborator saya, Jagriti Agrawal, yang sebenarnya telah menulis kode yang mengendalikan Mars rover saat ini dan juga membantu saya dalam menjelaskan serta mengembangkan contoh ini. Mari kita lihat lebih dekat.

## Contoh Pembelajaran Penguatan

Kita akan mengembangkan pembelajaran penguatan menggunakan contoh sederhana yang terinspirasi oleh Mars rover. Dalam aplikasi ini, rover dapat berada dalam salah satu dari enam posisi, yang ditunjukkan oleh enam kotak berikut. Misalkan rover mulai dari kotak keempat. Posisi Mars rover disebut **state** dalam pembelajaran penguatan, dan saya akan menyebut enam state ini sebagai **state 1**, **state 2**, **state 3**, **state 4**, **state 5**, dan **state 6**, dengan rover mulai dari **state 4**.

### Misi Sains

Rover dikirim ke Mars untuk melakukan berbagai misi sains. Rover dapat pergi ke berbagai tempat untuk menggunakan sensor seperti bor, radar, atau spektrometer untuk menganalisis batuan di berbagai lokasi di planet tersebut, atau mengambil foto menarik untuk ilmuwan di Bumi. Dalam contoh ini, **state 1** di sebelah kiri memiliki permukaan yang sangat menarik yang sangat diinginkan oleh para ilmuwan untuk diambil sampel. **State 6** juga memiliki permukaan yang cukup menarik, tetapi tidak sebanding dengan **state 1**. Kita lebih mungkin untuk melaksanakan misi sains di **state 1** daripada di **state 6**, meskipun **state 1** lebih jauh.

### Fungsi Reward

Cara kita merefleksikan bahwa **state 1** mungkin lebih berharga adalah melalui fungsi reward. Reward di **state 1** adalah 100, dan reward di **state 6** adalah 40. Reward di semua state lain di antara ini saya tulis sebagai reward nol karena tidak ada sains menarik yang dapat dilakukan di **state 2**, **state 3**, **state 4**, dan **state 5**.

Pada setiap langkah, rover mendapatkan pilihan satu dari dua aksi: bergerak ke kiri atau bergerak ke kanan. Pertanyaannya adalah, apa yang harus dilakukan rover? Dalam pembelajaran penguatan, kita sangat memperhatikan reward karena itu adalah cara kita mengetahui apakah robot melakukan dengan baik atau buruk.

## Contoh Aksi Rover

Mari kita lihat beberapa contoh tentang apa yang mungkin terjadi jika robot bergerak ke kiri, dimulai dari **state 4**. 

1. **Dari state 4**: Rover mendapatkan reward nol.
2. **Ke state 3**: Rover kembali mendapatkan reward nol.
3. **Ke state 2**: Reward tetap nol.
4. **Ke state 1**: Rover mendapatkan reward 100.

Untuk aplikasi ini, saya akan mengasumsikan bahwa ketika rover mencapai **state 1** atau **state 6**, hari berakhir. Dalam pembelajaran penguatan, kita kadang-kadang menyebut ini sebagai **terminal state**. Ini berarti bahwa setelah rover mencapai salah satu terminal state ini dan mendapatkan reward, tidak ada yang terjadi setelah itu. Mungkin robot kehabisan bahan bakar atau waktu habis untuk hari itu, yang menyebabkan rover hanya menikmati reward 100 atau 40, tetapi setelah itu tidak ada reward tambahan yang diperoleh.

Sebaliknya, rover juga dapat memilih untuk bergerak ke kanan. Dalam hal ini, dari **state 4**, rover akan mendapatkan:

1. **Reward di state 4**: Nol.
2. **Ke state 5**: Reward tetap nol.
3. **Ke state 6**: Rover mendapatkan reward 40.

Namun, bergerak ke kiri dan ke kanan adalah satu-satunya opsi. Salah satu hal yang dapat dilakukan rover adalah memulai dari **state 4** dan memutuskan untuk bergerak ke kanan. Setelah itu, rover mungkin mengubah pikirannya dan memutuskan untuk bergerak ke kiri sebagai berikut:

1. **Dari state 4**: Reward nol.
2. **Ke state 3**: Reward nol.
3. **Ke state 2**: Reward nol.
4. **Ke state 1**: Rover mendapatkan reward 100.

Dalam urutan aksi dan state ini, robot membuang sedikit waktu. Jadi, mungkin ini bukan cara terbaik untuk mengambil aksi, tetapi ini adalah salah satu pilihan yang dapat dipilih oleh algoritma, meskipun harapannya Anda tidak memilih opsi ini.

## Ringkasan

Untuk merangkum, pada setiap langkah waktu, robot berada dalam suatu **state** yang akan saya sebut **S**, dan ia mendapatkan pilihan aksi serta menikmati beberapa reward, **R(S)**, yang diperoleh dari state tersebut. Akibat dari aksi ini, ia pindah ke state baru **S'**. Sebagai contoh konkret, ketika robot berada di **state 4** dan mengambil aksi untuk pergi ke kiri, ia mungkin tidak menikmati reward nol yang terkait dengan **state 4** dan tidak akan memiliki state baru **state 3**. 

Ketika Anda mempelajari algoritma pembelajaran penguatan spesifik, Anda akan melihat bahwa keempat hal ini  state, aksi, reward, dan next state  adalah elemen inti yang akan diperhatikan oleh algoritma pembelajaran penguatan saat memutuskan bagaimana mengambil aksi.



### Kejelasan Reward

Untuk kejelasan, reward di sini, **R(S)**, adalah reward yang terkait dengan state ini. Reward nol terkait dengan **state 4** daripada dengan **state 3**. Itulah formalism bagaimana aplikasi pembelajaran penguatan bekerja.
