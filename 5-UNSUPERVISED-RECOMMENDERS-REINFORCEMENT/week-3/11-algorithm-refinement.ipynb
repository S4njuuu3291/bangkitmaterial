# Arsitektur Jaringan Syaraf untuk Q-Learning

Dalam video terakhir, kita melihat arsitektur **neural network** yang akan menerima **state** dan **action**, kemudian mencoba mengeluarkan fungsi **Q**, yaitu \( Q(s, a) \). Ternyata, ada perubahan pada arsitektur neural network yang membuat algoritma ini jauh lebih efisien. Sebagian besar implementasi **DQN** sebenarnya menggunakan arsitektur yang lebih efisien ini, yang akan kita lihat dalam video ini. Mari kita lihat.

## Arsitektur Neural Network Sebelumnya

Ini adalah arsitektur neural network yang kita lihat sebelumnya, di mana input terdiri dari 12 angka dan outputnya adalah \( Q(s, a) \). 

- Setiap kali kita berada dalam suatu **state** \( s \), kita harus melakukan inferensi pada neural network secara terpisah sebanyak empat kali untuk menghitung keempat nilai ini agar kita dapat memilih **action** \( a \) yang memberikan nilai **Q** terbesar. 
- Ini tidak efisien karena kita harus melakukan inferensi empat kali dari setiap **state**.

## Arsitektur Neural Network yang Dimodifikasi

Ternyata lebih efisien untuk melatih satu neural network yang dapat mengeluarkan semua empat nilai ini secara bersamaan. Inilah tampilan arsitekturnya:

- Input terdiri dari **delapan angka** yang sesuai dengan **state** dari lunar lander.
- Neural network melalui dua **hidden layer** dengan masing-masing 64 unit.
- Outputnya memiliki empat unit output yang bertugas menghitung \( Q(s, a) \) untuk semua tindakan yang mungkin:

  - \( Q(s, \text{nothing}) \)
  - \( Q(s, \text{left}) \)
  - \( Q(s, \text{main}) \)
  - \( Q(s, \text{right}) \)

## Efisiensi yang Dihasilkan

Dengan menggunakan arsitektur ini, kita hanya perlu menjalankan inferensi sekali untuk mendapatkan semua empat nilai tersebut. Kemudian, kita dapat dengan cepat memilih **action** \( a \) yang memaksimalkan \( Q(s, a) \). 

Anda juga akan melihat bahwa dalam **persamaan Bellman**, ada langkah di mana kita harus menghitung:

\[
\max_{a'} Q(s', a') \cdot \gamma + R(s)
\]

Jaringan neural ini juga membuatnya jauh lebih efisien untuk menghitung nilai ini, karena kita mendapatkan \( Q(s', a') \) untuk semua **actions** \( a' \) pada saat yang bersamaan. Dengan begitu, kita hanya perlu memilih nilai maksimum untuk menghitung nilai di sisi kanan persamaan Bellman.
