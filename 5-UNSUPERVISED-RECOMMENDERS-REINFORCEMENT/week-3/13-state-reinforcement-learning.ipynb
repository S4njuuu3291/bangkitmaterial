# Reinforcement Learning

Reinforcement learning adalah **seperangkat teknologi** yang menarik. Sebenarnya, saat saya mengerjakan tesis PhD saya, reinforcement learning adalah subjek utama tesis saya. Jadi, saya sangat bersemangat tentang ide-ide ini. 

Namun, meskipun ada banyak momentum penelitian dan kegembiraan di balik reinforcement learning, saya merasa ada sedikit, atau mungkin seringkali banyak, **hype** di sekitarnya. Oleh karena itu, saya berharap dapat berbagi dengan Anda pemahaman praktis tentang di mana posisi reinforcement learning saat ini dalam hal **utilitas untuk aplikasi**.

## Hype di Balik Reinforcement Learning

Salah satu alasan di balik hype tentang reinforcement learning adalah bahwa banyak publikasi penelitian berfokus pada **lingkungan simulasi**. Setelah bekerja di kedua bidang, yaitu simulasi dan robot nyata, saya dapat memberitahu Anda bahwa jauh lebih mudah untuk membuat algoritma reinforcement learning berfungsi dalam simulasi atau video game dibandingkan dengan robot nyata. 

Banyak pengembang telah mengomentari bahwa meskipun mereka berhasil membuatnya berfungsi di simulasi, ternyata menjadi tantangan yang mengejutkan untuk menerapkannya di dunia nyata atau pada robot nyata. Oleh karena itu, jika Anda menerapkan algoritma ini pada aplikasi nyata, ini adalah salah satu keterbatasan yang perlu Anda perhatikan untuk memastikan bahwa apa yang Anda lakukan dapat berfungsi di aplikasi nyata.

## Aplikasi Reinforcement Learning

Kedua, meskipun ada banyak liputan media tentang reinforcement learning, saat ini ada jauh lebih sedikit aplikasi reinforcement learning dibandingkan dengan **supervised learning** dan **unsupervised learning**. Jadi, jika Anda membangun aplikasi praktis, kemungkinan besar Anda akan menemukan bahwa supervised learning atau unsupervised learning lebih berguna atau merupakan alat yang tepat untuk pekerjaan tersebut, dibandingkan dengan menggunakan reinforcement learning.

Saya sendiri telah menggunakan reinforcement learning beberapa kali, terutama dalam aplikasi **kendali robot**, tetapi dalam pekerjaan saya sehari-hari, saya lebih sering menggunakan supervised dan unsupervised learning. 

## Potensi Reinforcement Learning

Ada banyak penelitian menarik dalam reinforcement learning saat ini, dan saya percaya bahwa potensi reinforcement learning untuk aplikasi di masa depan sangat besar. Reinforcement learning masih tetap menjadi salah satu pilar utama dari machine learning. Dengan memiliki kerangka kerja ini saat Anda mengembangkan algoritma machine learning Anda sendiri, saya berharap ini akan membuat Anda lebih efektif dalam membangun sistem machine learning yang bekerja dengan baik.
