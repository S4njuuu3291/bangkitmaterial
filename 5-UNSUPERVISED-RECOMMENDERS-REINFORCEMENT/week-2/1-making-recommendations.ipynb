## Contoh Studi Kasus: Prediksi Rating Film

Sebagai contoh, misalkan Anda menjalankan sebuah situs streaming film besar dan pengguna Anda memberikan rating film dengan skala 1 hingga 5 bintang. Dalam sistem rekomendasi, Anda memiliki satu set pengguna—katakanlah ada empat pengguna: **Alice**, **Bob**, **Carol**, dan **Dave**. Setiap pengguna memiliki ID numerik: 1, 2, 3, dan 4.

Selain itu, Anda memiliki satu set film, misalnya: **Love at Last**, **Romance Forever**, **Cute Puppies of Love**, **Nonstop Car Chases**, dan **Sword vs. Karate**. Pengguna-pengguna ini telah memberikan rating pada film-film tersebut, mulai dari 1 hingga 5 bintang.

Namun, beberapa film mungkin belum mereka tonton, jadi Anda tidak memiliki rating untuk itu. Misalnya, Alice memberikan rating 5 bintang untuk **Love at Last** dan **Romance Forever**, tapi belum menonton **Cute Puppies of Love**, jadi kita akan tandai itu dengan tanda tanya (?). Untuk **Nonstop Car Chases** dan **Sword vs. Karate**, dia memberikan rating 0 bintang.



## Menentukan Notasi dalam Recommender Systems

Dalam **recommender system**, Anda memiliki sejumlah pengguna dan sejumlah item (dalam hal ini, film) yang ingin Anda rekomendasikan kepada pengguna.

- Saya akan menggunakan notasi **nu** untuk menunjukkan jumlah pengguna. Pada contoh ini, **nu = 4** (karena ada 4 pengguna).
- Saya juga akan menggunakan notasi **nm** untuk menunjukkan jumlah film (item). Pada contoh ini, **nm = 5** (karena ada 5 film).

### Definisi Variabel:
1. **r(i,j)** = 1, jika pengguna **j** telah memberikan rating untuk film **i**.
   - Contoh: **r(1,1) = 1** karena Alice (pengguna 1) telah memberikan rating untuk film pertama. Namun, **r(3,1) = 0** karena dia belum menonton film ketiga.

2. **y(i,j)** digunakan untuk menunjukkan rating yang diberikan oleh pengguna **j** pada film **i**.
   - Contoh: Rating untuk film ketiga yang diberikan oleh pengguna kedua adalah 4. Jadi, **y(3,2) = 4**.

### Langkah Berikutnya

Langkah berikutnya adalah mencoba memprediksi bagaimana pengguna akan memberi rating pada film yang belum mereka tonton. Dengan demikian, kita dapat merekomendasikan item yang kemungkinan besar akan mereka beri rating 5 bintang. Pada video berikutnya, kita akan mulai mengembangkan algoritma untuk melakukan hal ini, dengan satu asumsi khusus: kita akan berasumsi bahwa kita memiliki akses ke fitur tambahan mengenai film, seperti apakah film tersebut bergenre **romance** atau **action**.

Nantinya, kita akan melihat bagaimana algoritma tersebut bekerja, bahkan jika kita tidak memiliki informasi tambahan ini. Sekarang, mari kita lanjut ke video berikutnya untuk mulai membangun algoritmanya.
