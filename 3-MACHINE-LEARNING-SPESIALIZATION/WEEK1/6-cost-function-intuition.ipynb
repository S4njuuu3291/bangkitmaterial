# Penjelasan tentang Fungsi Biaya dalam Regresi Linear

Kita sedang melihat **definisi matematis dari fungsi biaya**. Sekarang, mari kita bangun sedikit pemahaman tentang apa yang sebenarnya dilakukan oleh fungsi biaya. Dalam video ini, kita akan berjalan melalui satu contoh untuk melihat bagaimana fungsi biaya dapat digunakan untuk menemukan parameter terbaik untuk model Anda. Saya tahu video ini sedikit lebih panjang daripada yang lain, tapi harap bersabar, saya rasa ini akan sepadan.

## Ringkasan Fungsi Biaya yang Telah Kita Lihat

Untuk merangkum, berikut ini yang telah kita lihat tentang fungsi biaya sejauh ini. Anda ingin menyesuaikan garis lurus dengan data pelatihan, sehingga Anda memiliki model sebagai berikut:

$$
f_{w,b}(x) = w \cdot x + b
$$

Di sini, parameter modelnya adalah \( w \) dan \( b \). Sekarang, tergantung pada nilai yang dipilih untuk parameter ini, Anda akan mendapatkan berbagai garis lurus seperti ini. Anda ingin menemukan nilai untuk \( w \) dan \( b \), sehingga garis lurus tersebut cocok dengan data pelatihan dengan baik.

Untuk mengukur seberapa baik pilihan \( w \) dan \( b \) cocok dengan data pelatihan, Anda memiliki **fungsi biaya \( J \)**. Apa yang dilakukan oleh fungsi biaya \( J \) adalah mengukur perbedaan antara prediksi model dan nilai sebenarnya dari \( y \).

Nantinya, kita akan melihat bahwa regresi linear akan mencoba menemukan nilai \( w \) dan \( b \), sehingga \( J(w, b) \) sekecil mungkin. Dalam matematika, kita menulisnya seperti ini:

$$
\text{Minimalkan } J(w, b)
$$

## Visualisasi Fungsi Biaya dengan Model Sederhana

Untuk memvisualisasikan fungsi biaya \( J \) dengan lebih baik, kita akan menggunakan versi sederhana dari model regresi linear. Kita akan menggunakan model:

$$
f_w(x) = w \cdot x
$$

Anda bisa menganggap ini sebagai mengambil model asli di sebelah kiri, dan menghilangkan parameter \( b \), atau menyetel \( b = 0 \). Jadi, sekarang \( f_w(x) = w \cdot x \) hanya memiliki satu parameter \( w \), dan fungsi biaya \( J \) terlihat mirip dengan sebelumnya, hanya saja sekarang \( f = w \cdot x \) dan \( J \) sekarang hanya sebagai fungsi dari \( w \).

Tujuannya menjadi sedikit berbeda juga, karena sekarang Anda hanya memiliki satu parameter, yaitu \( w \), bukan \( w \) dan \( b \). Dengan model sederhana ini, tujuannya adalah menemukan nilai untuk \( w \) yang meminimalkan \( J(w) \).

## Perbandingan Grafis \( f_w(x) \) dan \( J(w) \)

Untuk melihat ini secara visual, mari kita lihat bagaimana fungsi biaya berubah saat Anda memilih berbagai nilai untuk parameter \( w \). Secara khusus, mari kita lihat grafik dari model \( f_w(x) \), dan fungsi biaya \( J(w) \). Saya akan plot grafik ini berdampingan, dan Anda akan bisa melihat bagaimana keduanya saling berhubungan.

### Grafik Fungsi Model \( f_w(x) \)

Perhatikan bahwa untuk \( f_w(x) \), ketika parameter \( w \) tetap, maka \( f_w \) hanya merupakan fungsi dari \( x \), yang berarti nilai \( y \) yang diperkirakan bergantung pada nilai \( x \). Sebaliknya, fungsi biaya \( J(w) \), adalah fungsi dari \( w \), di mana \( w \) mengontrol kemiringan garis yang didefinisikan oleh \( f_w(x) \).

### Fungsi Biaya \( J(w) \)

Fungsi biaya \( J \) bergantung pada parameter \( w \). Mari kita plot fungsi \( f_w(x) \) di sebelah kiri dan \( J(w) \) di sebelah kanan agar Anda bisa melihat bagaimana keduanya saling berhubungan.

#### Contoh dengan \( w = 1 \)

Misalkan kita memilih \( w = 1 \). Untuk pilihan \( w = 1 \), fungsi \( f_w(x) \) akan menjadi garis lurus dengan kemiringan 1. Selanjutnya, kita dapat menghitung fungsi biaya \( J \) ketika \( w = 1 \).

Fungsi biaya didefinisikan sebagai berikut, yaitu fungsi **error kuadrat**:

$$
J(w) = \frac{1}{2m} \sum_{i=1}^{m} (f_w(x^i) - y^i)^2
$$

Di mana $$ f_w(x^i) = w \cdot x^i \ dan \ y^i $$ adalah nilai aktual. Jika \( w = 1 \), maka perbedaan \( f_w(x^i) - y^i \) adalah 0 untuk setiap titik data. Oleh karena itu, fungsi biaya \( J(1) = 0 \).

### Contoh dengan \( w = 0.5 \)

Sekarang, jika \( w = 0.5 \), maka \( f_w(x) \) akan menjadi garis dengan kemiringan 0.5. Untuk menghitung fungsi biaya \( J \), kita gunakan rumus yang sama, menghitung selisih kuadrat antara nilai estimasi \( f_w(x) \) dan nilai aktual \( y \).

Untuk setiap titik data, kita hitung error kuadrat dan jumlahkan semuanya, kemudian bagi dengan \( 2m \). Dalam kasus ini, hasilnya adalah sekitar \( 0.58 \).

### Contoh dengan \( w = 0 \)

Jika \( w = 0 \), maka \( f_w(x) \) menjadi garis horizontal di sumbu \( x \). Fungsi biaya \( J \) ketika \( w = 0 \) adalah \( 2.33 \).

## Kesimpulan

Dari berbagai nilai \( w \) yang diuji, kita dapat melihat bagaimana **fungsi biaya** \( J(w) \) berubah dan mencari nilai \( w \) yang **meminimalkan** fungsi biaya tersebut. Dalam regresi linear, kita ingin menemukan nilai \( w \) yang meminimalkan \( J(w) \) dan memberikan **fit terbaik** untuk data pelatihan.
