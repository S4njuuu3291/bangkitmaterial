  ## Pembelajaran Terawasi dan Aplikasinya

Pembelajaran mesin, terutama **pembelajaran terawasi (supervised learning)**, saat ini menciptakan dampak ekonomi yang signifikan, dengan sekitar **99% nilai** yang dihasilkan oleh pembelajaran terawasi. Dalam pembelajaran terawasi, algoritma belajar dari contoh data **input (x)** yang dipasangkan dengan **output (y)**, yang disebut **label**. Dengan melihat pasangan input-output ini, algoritma belajar memetakan input ke output, sehingga mampu memprediksi hasil yang tepat untuk data baru yang belum pernah dilihat sebelumnya.

### Contoh Aplikasi Pembelajaran Terawasi:
- **Filter spam**: Memprediksi apakah email tertentu adalah spam atau bukan.
- **Pengenalan suara**: Mengubah klip audio menjadi teks.
- **Terjemahan mesin**: Menerjemahkan teks dari satu bahasa ke bahasa lain.
- **Iklan online**: Memrediksi kemungkinan pengguna mengklik iklan, membantu platform besar menghasilkan pendapatan melalui klik iklan.
- **Mobil tanpa pengemudi**: Algoritma mendeteksi posisi kendaraan lain berdasarkan data sensor.
- **Inspeksi visual**: Deteksi cacat produk seperti goresan atau penyok di industri manufaktur.

### Dua Jenis Masalah Pembelajaran Terawasi:
1. **Regresi**: Memprediksi nilai numerik, misalnya harga rumah berdasarkan ukurannya.
2. **Klasifikasi**: Memprediksi kategori, misalnya menentukan apakah email adalah spam atau tidak.

Model pembelajaran terawasi dilatih dengan data berlabel, kemudian digunakan untuk menghasilkan prediksi yang akurat pada data baru. Contoh regresi melibatkan prediksi **harga rumah**, di mana algoritma dapat menyesuaikan **garis lurus** atau **kurva** untuk memprediksi harga berdasarkan data ukuran rumah. Pilihan metode regresi atau klasifikasi tergantung pada jenis data dan tujuan prediksi.


Algoritma pembelajaran terawasi (supervised learning) belajar untuk memprediksi hubungan antara input dan output (X ke Y). Di video sebelumnya, Anda telah melihat bahwa algoritma **regresi** (sejenis algoritma pembelajaran terawasi) belajar untuk memprediksi angka dari sekian banyak angka yang tak terbatas.

Ada jenis utama kedua dari algoritma pembelajaran terawasi, yaitu **algoritma klasifikasi**. Mari kita lihat contoh masalah klasifikasi.

## Contoh: Deteksi Kanker Payudara
Misalnya, Anda sedang membangun sistem pembelajaran mesin untuk membantu dokter mendeteksi kanker payudara. Deteksi dini kanker sangat penting karena bisa menyelamatkan nyawa pasien. Sistem ini menggunakan catatan medis pasien untuk memprediksi apakah tumor (benjolan) itu **ganas** (kanker) atau **jinak** (bukan kanker).

Dataset Anda mungkin berisi data tumor dengan berbagai ukuran. Tumor-tumor ini dilabeli sebagai **jinak** (ditandai dengan angka 0) atau **ganas** (ditandai dengan angka 1). 

- Sumbu horizontal mewakili **ukuran tumor**.
- Sumbu vertikal hanya memiliki dua nilai: **0 (jinak)** atau **1 (ganas)**.

### Perbedaan dengan Regresi
Yang membedakan masalah ini dengan **regresi** adalah dalam klasifikasi, kita hanya memprediksi sejumlah kecil hasil atau kategori (dalam hal ini, 0 atau 1). Sedangkan regresi mencoba memprediksi **angka dari banyak kemungkinan yang tak terbatas**.

## Visualisasi Data
Data bisa dipetakan pada grafik seperti ini:
- **Lingkaran (O)**: Menunjukkan tumor yang jinak (0).
- **Tanda silang (X)**: Menunjukkan tumor yang ganas (1).

Jika seorang pasien baru datang dengan tumor yang ukuran tertentu, kita ingin tahu apakah tumor tersebut jinak atau ganas. Algoritma klasifikasi akan mencari **batasan** yang memisahkan tumor jinak dan ganas. Batasan ini membantu dokter membuat keputusan diagnosis.

## Lebih dari Dua Kategori Keluaran
Masalah klasifikasi tidak harus hanya memiliki dua kategori. Misalnya, jika tumor tersebut ganas, algoritma bisa memprediksi jenis kanker tertentu (misalnya tipe 1 atau tipe 2). Dalam kasus ini, algoritma harus memprediksi lebih dari dua kategori hasil.

## Apa itu Klasifikasi?
Secara umum, algoritma klasifikasi memprediksi kategori, dan kategori tidak selalu berupa angka. Misalnya:
- Memprediksi apakah gambar adalah **kucing** atau **anjing**.
- Memprediksi apakah tumor **jinak** atau **ganas**.

Kategori juga bisa berupa angka seperti **0, 1** atau **0, 1, 2**. Yang membedakan klasifikasi dari regresi adalah:
- **Klasifikasi** memprediksi **beberapa kategori terbatas** (seperti 0, 1, atau 2).
- **Regresi** memprediksi **angka tak terbatas** (misalnya 0.5, 1.7).

## Menambahkan Input Lebih Banyak
Di contoh sebelumnya, kita hanya menggunakan satu input: **ukuran tumor**. Namun, kita juga bisa menggunakan lebih dari satu input untuk memprediksi hasil. 

Misalnya, Anda juga memiliki **usia pasien** sebagai input tambahan. Sekarang dataset Anda memiliki dua input: **usia pasien** dan **ukuran tumor**.

Pada dataset ini:
- **Lingkaran (O)** menunjukkan pasien dengan tumor jinak.
- **Tanda silang (X)** menunjukkan pasien dengan tumor ganas.

Dengan menggunakan dua input ini, algoritma pembelajaran mesin akan mencari **batasan** untuk memisahkan tumor jinak dari yang ganas.

## Kesimpulan
- **Pembelajaran terawasi** memetakan input (X) ke output (Y), di mana algoritma belajar dari **jawaban yang benar**.
- Ada dua jenis utama dari pembelajaran terawasi: **Regresi** dan **Klasifikasi**.
  - **Regresi**: Memprediksi angka dari banyak kemungkinan.
  - **Klasifikasi**: Memprediksi kategori dari sejumlah kecil kategori.

Jadi, sekarang Anda sudah tahu apa itu pembelajaran terawasi, termasuk regresi dan klasifikasi.
