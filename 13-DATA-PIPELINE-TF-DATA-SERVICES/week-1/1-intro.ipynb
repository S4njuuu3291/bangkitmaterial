## Pendahuluan

Ketika kita memikirkan *machine learning*, kita biasanya hanya memikirkan model-model hebat yang sekarang dapat kita buat. Namun, sebelum Anda lebih jauh membangun model, **data** adalah elemen penting dari setiap model. Kursus ini akan membantu Anda dalam mengatur, menstrukturkan, dan mengalirkan data Anda, serta memperkenalkan berbagai sumber data yang dapat Anda gunakan.

## Jenis Data dalam Machine Learning

Sebagai langkah awal, kita akan melihat jenis-jenis data yang biasanya ditemui dalam *machine learning*. Seiring waktu, kami menyadari bahwa tipe data yang digunakan untuk membuat model dan juga untuk **inference** cenderung terbagi ke dalam kategori-kategori utama berikut:

### 1. **Gambar (Image)**

   Data gambar digunakan untuk **computer vision**, seperti pengenalan objek atau klasifikasi gambar.

### 2. **Audio**

   Digunakan untuk pengenalan suara atau musik. Misalnya, "Hey Google, lagu apa ini?"



### 3. **Data Terstruktur**

   Data terstruktur sering digunakan untuk masalah seperti **regresi** atau **rekomendasi**.

### 4. **Video**

   Video mirip dengan gambar, namun digunakan untuk mengenali, melacak, atau memahami konten dalam gambar yang bergerak.

### 5. **Teks**

   Data teks digunakan untuk **klasifikasi**, **analisis sentimen**, **terjemahan**, dan lainnya. Terjemahan, misalnya, sangat berguna untuk membantu kita lebih memahami dunia di sekitar kita.

---

Terjemahan adalah salah satu hal yang sangat bermanfaat untuk membantu kita memahami lebih baik tentang dunia di sekitar kita.
